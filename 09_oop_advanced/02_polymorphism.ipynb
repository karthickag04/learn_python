## 1. Method Overriding Polymorphism

```python
# Same method name, different behavior

class Shape:
    def area(self):
        raise NotImplementedError("Subclass must implement")
    
    def perimeter(self):
        raise NotImplementedError("Subclass must implement")

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        import math
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        import math
        return 2 * math.pi * self.radius

class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c
    
    def area(self):
        import math
        s = self.perimeter() / 2
        return math.sqrt(s * (s-self.a) * (s-self.b) * (s-self.c))
    
    def perimeter(self):
        return self.a + self.b + self.c

# Polymorphic function - works with ANY shape
def print_shape_info(shape):
    print(f"{shape.__class__.__name__}:")
    print(f"  Area: {shape.area():.2f}")
    print(f"  Perimeter: {shape.perimeter():.2f}")

# Test with different shapes
shapes = [
    Rectangle(10, 5),
    Circle(7),
    Triangle(3, 4, 5)
]

for shape in shapes:
    print_shape_info(shape)
    print()
```

## 2. Duck Typing

```python
# "If it walks like a duck and quacks like a duck, it's a duck"
# No inheritance needed - just implement the methods

class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

class Robot:
    def speak(self):
        return "Beep boop!"

class Human:
    def speak(self):
        return "Hello!"

# Function doesn't care about type, just that it has speak()
def make_speak(thing):
    return thing.speak()

# No common parent needed!
for obj in [Dog(), Cat(), Robot(), Human()]:
    print(f"{obj.__class__.__name__} says: {make_speak(obj)}")
```

```python
# Practical duck typing: File-like objects

class StringBuffer:
    """Acts like a file but stores in memory"""
    def __init__(self):
        self.data = ""
    
    def write(self, text):
        self.data += text
    
    def read(self):
        return self.data

class Logger:
    """Logs with timestamp"""
    def __init__(self):
        self.entries = []
    
    def write(self, text):
        from datetime import datetime
        self.entries.append(f"[{datetime.now():%H:%M:%S}] {text}")
    
    def read(self):
        return "\n".join(self.entries)

# Function works with anything that has write()
def log_message(writer, message):
    writer.write(message)

# Test with different "writers"
buffer = StringBuffer()
logger = Logger()

log_message(buffer, "Hello ")
log_message(buffer, "World!")
print(f"Buffer: {buffer.read()}")

log_message(logger, "User logged in")
log_message(logger, "Action performed")
print(f"\nLogger:\n{logger.read()}")
```

## 3. Operator Overloading

```python
# Make custom objects work with operators

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):  # +
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):  # -
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):  # *
        return Vector(self.x * scalar, self.y * scalar)
    
    def __rmul__(self, scalar):  # scalar * vector
        return self.__mul__(scalar)
    
    def __eq__(self, other):  # ==
        return self.x == other.x and self.y == other.y
    
    def __abs__(self):  # abs()
        import math
        return math.sqrt(self.x**2 + self.y**2)
    
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

# Test operators
v1 = Vector(3, 4)
v2 = Vector(1, 2)

print(f"v1 = {v1}")
print(f"v2 = {v2}")
print(f"v1 + v2 = {v1 + v2}")
print(f"v1 - v2 = {v1 - v2}")
print(f"v1 * 2 = {v1 * 2}")
print(f"3 * v2 = {3 * v2}")
print(f"|v1| = {abs(v1)}")
print(f"v1 == v1? {v1 == Vector(3, 4)}")
```

```python
# Money class with operators

class Money:
    def __init__(self, amount, currency="USD"):
        self.amount = round(amount, 2)
        self.currency = currency
    
    def __add__(self, other):
        if self.currency != other.currency:
            raise ValueError("Cannot add different currencies")
        return Money(self.amount + other.amount, self.currency)
    
    def __sub__(self, other):
        if self.currency != other.currency:
            raise ValueError("Cannot subtract different currencies")
        return Money(self.amount - other.amount, self.currency)
    
    def __mul__(self, factor):
        return Money(self.amount * factor, self.currency)
    
    def __truediv__(self, factor):
        return Money(self.amount / factor, self.currency)
    
    def __lt__(self, other):
        if self.currency != other.currency:
            raise ValueError("Cannot compare different currencies")
        return self.amount < other.amount
    
    def __eq__(self, other):
        return self.amount == other.amount and self.currency == other.currency
    
    def __repr__(self):
        return f"{self.currency} {self.amount:,.2f}"

# Test
price = Money(99.99)
tax = Money(8.50)
discount = Money(15.00)

print(f"Price: {price}")
print(f"Tax: {tax}")
print(f"Total: {price + tax}")
print(f"After discount: {price + tax - discount}")
print(f"Double price: {price * 2}")
print(f"Split 3 ways: {price / 3}")

# Comparison
print(f"\nPrice > Tax? {price > tax}")
```

```python
# Common special methods for operator overloading

# Comparison operators
print("Comparison Operators:")
operators_comparison = {
    '__lt__': '< (less than)',
    '__le__': '<= (less or equal)',
    '__gt__': '> (greater than)',
    '__ge__': '>= (greater or equal)',
    '__eq__': '== (equal)',
    '__ne__': '!= (not equal)'
}
for method, op in operators_comparison.items():
    print(f"  {method}: {op}")

# Arithmetic operators
print("\nArithmetic Operators:")
operators_arithmetic = {
    '__add__': '+ (add)',
    '__sub__': '- (subtract)',
    '__mul__': '* (multiply)',
    '__truediv__': '/ (divide)',
    '__floordiv__': '// (floor divide)',
    '__mod__': '% (modulo)',
    '__pow__': '** (power)'
}
for method, op in operators_arithmetic.items():
    print(f"  {method}: {op}")
```

## 4. Complete Example: Payment System

```python
# Polymorphic payment processing

class PaymentMethod:
    """Base payment class"""
    def process(self, amount):
        raise NotImplementedError
    
    def validate(self):
        raise NotImplementedError

class CreditCard(PaymentMethod):
    def __init__(self, card_number, expiry, cvv):
        self.card_number = card_number
        self.expiry = expiry
        self.cvv = cvv
    
    def validate(self):
        # Simple validation
        if len(self.card_number.replace(' ', '')) != 16:
            return False, "Invalid card number"
        if len(self.cvv) != 3:
            return False, "Invalid CVV"
        return True, "Card validated"
    
    def process(self, amount):
        valid, msg = self.validate()
        if not valid:
            return False, msg
        
        last_four = self.card_number[-4:]
        return True, f"Charged ${amount:.2f} to card ending in {last_four}"

class PayPal(PaymentMethod):
    def __init__(self, email):
        self.email = email
    
    def validate(self):
        if '@' not in self.email:
            return False, "Invalid email"
        return True, "PayPal account validated"
    
    def process(self, amount):
        valid, msg = self.validate()
        if not valid:
            return False, msg
        return True, f"Charged ${amount:.2f} to PayPal ({self.email})"

class BankTransfer(PaymentMethod):
    def __init__(self, routing, account):
        self.routing = routing
        self.account = account
    
    def validate(self):
        if len(self.routing) != 9:
            return False, "Invalid routing number"
        return True, "Bank account validated"
    
    def process(self, amount):
        valid, msg = self.validate()
        if not valid:
            return False, msg
        return True, f"Transferred ${amount:.2f} from bank (****{self.account[-4:]})"

class Cryptocurrency(PaymentMethod):
    RATES = {'BTC': 43000, 'ETH': 2300}  # USD per coin
    
    def __init__(self, wallet, currency='BTC'):
        self.wallet = wallet
        self.currency = currency
    
    def validate(self):
        if len(self.wallet) < 26:
            return False, "Invalid wallet address"
        return True, "Wallet validated"
    
    def process(self, amount):
        valid, msg = self.validate()
        if not valid:
            return False, msg
        
        rate = self.RATES.get(self.currency, 1)
        crypto_amount = amount / rate
        return True, f"Sent {crypto_amount:.6f} {self.currency} (${amount:.2f})"

# Polymorphic checkout function
def checkout(payment_method, amount):
    """Process payment with any payment method"""
    print(f"\nProcessing ${amount:.2f} payment...")
    success, message = payment_method.process(amount)
    status = "âœ…" if success else "âŒ"
    print(f"{status} {message}")
    return success

# Demo
print("ðŸ’³ PAYMENT SYSTEM DEMO")
print("=" * 50)

# Different payment methods
payments = [
    (CreditCard("4111 1111 1111 1111", "12/25", "123"), 99.99),
    (PayPal("user@example.com"), 49.99),
    (BankTransfer("123456789", "0001234567"), 500.00),
    (Cryptocurrency("1BvBMSEYstWetqTFn5Au4m4GFg7xJaNVN2", "BTC"), 1000.00)
]

for method, amount in payments:
    checkout(method, amount)
```

## Summary

### Polymorphism Types:

| Type | Description | Example |
|------|-------------|----------|
| Method Override | Same method, different behavior | `shape.area()` |
| Duck Typing | No inheritance needed | `thing.speak()` |
| Operator Overload | Custom operators | `v1 + v2` |

### Key Operators to Overload:

| Method | Operator | Use |
|--------|----------|-----|
| `__add__` | `+` | Addition |
| `__sub__` | `-` | Subtraction |
| `__eq__` | `==` | Equality |
| `__lt__` | `<` | Less than |
| `__str__` | `str()` | String conversion |

### Benefits:
1. Write generic code that works with many types
2. Easy to extend without modifying existing code
3. Natural syntax with operator overloading
4. Flexible duck typing for Python
