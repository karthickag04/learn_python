## 1. Basic Inheritance

````python
class TrainingCenter01:
    def __init__(self, name, location):
        self.name = name
        self.location = location
        self.courses = ["Python Basics", "Data Science 101", "Web Development"]
        self.trainers = ["Alice", "Bob", "Charlie"]

    def add_course(self, course):
        self.courses.append(course)
    
    def list_of_trainers(self):
        return self.trainers

    def list_courses(self):
        return self.courses
    
class TrainingCenter02:
    def __init__(self, name, location):
        self.name = name
        self.location = location
        self.courses = ["Python ", "Java", "Android Development"]
        self.trainers = ["Ram", "Shyam", "Geeta"]

    def add_course(self, course):
        self.courses.append(course)
    
    def list_of_trainers(self):
        return self.trainers

    def list_courses(self):
        return self.courses
````

````python
tc = TrainingCenter01("Tech Academy", "New York")
tc.add_course("Machine Learning")
print(tc.list_courses())
````

````python
tc = TrainingCenter02("Edu Tech", "New York")
tc.add_course("MERN Stack Development")
print(tc.list_courses())
````

````python
# Parent class (Base class)
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def speak(self):
        return "Some sound"
    
    def describe(self):
        return f"{self.name} is {self.age} years old"

# Child class (Derived class)
class Dog(Animal):  # Dog inherits from Animal
    def speak(self):  # Override parent method
        return "Woof!"
    
    def fetch(self):  # New method only for Dog
        return f"{self.name} fetches the ball!"

class Cat(Animal):
    def speak(self):
        return "Meow!"
    
    def scratch(self):
        return f"{self.name} scratches the furniture!"

# Use the classes
dog = Dog("Buddy", 3)
cat = Cat("Whiskers", 5)

# Inherited methods work
print(dog.describe())  # From Animal
print(cat.describe())  # From Animal

# Overridden methods
print(f"Dog says: {dog.speak()}")
print(f"Cat says: {cat.speak()}")

# Unique methods
print(dog.fetch())
print(cat.scratch())
````

````python
# Check inheritance relationships
print(f"Dog is Animal? {isinstance(dog, Animal)}")
print(f"Cat is Animal? {isinstance(cat, Animal)}")
print(f"Dog is Cat? {isinstance(dog, Cat)}")

# Check class hierarchy
print(f"\nDog's parent classes: {Dog.__bases__}")
print(f"Dog is subclass of Animal? {issubclass(Dog, Animal)}")
````

## 2. The `super()` Function

````python
# super() calls the parent class method

class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def info(self):
        return f"{self.brand} {self.model}"

class Car(Vehicle):
    def __init__(self, brand, model, doors):
        super().__init__(brand, model)  # Call parent __init__
        self.doors = doors
    
    def info(self):
        # Call parent info() and extend it
        return f"{super().info()} ({self.doors} doors)"

class ElectricCar(Car):
    def __init__(self, brand, model, doors, battery_capacity):
        super().__init__(brand, model, doors)
        self.battery_capacity = battery_capacity
    
    def info(self):
        return f"{super().info()} - {self.battery_capacity} kWh battery"

# Test
vehicle = Vehicle("Generic", "Transport")
car = Car("Toyota", "Camry", 4)
tesla = ElectricCar("Tesla", "Model 3", 4, 75)

print(vehicle.info())
print(car.info())
print(tesla.info())
````

````python
# Real-world example: Employee hierarchy

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.base_salary = salary
    
    def get_salary(self):
        return self.base_salary
    
    def __str__(self):
        return f"{self.name}: ${self.get_salary():,.2f}/year"

class Manager(Employee):
    def __init__(self, name, salary, bonus):
        super().__init__(name, salary)
        self.bonus = bonus
    
    def get_salary(self):
        return super().get_salary() + self.bonus

class Developer(Employee):
    def __init__(self, name, salary, language):
        super().__init__(name, salary)
        self.language = language
    
    def __str__(self):
        return f"{super().__str__()} ({self.language} Developer)"

class SeniorDeveloper(Developer):
    def __init__(self, name, salary, language, team_size):
        super().__init__(name, salary, language)
        self.team_size = team_size
    
    def get_salary(self):
        # 10% extra per team member
        return super().get_salary() * (1 + 0.1 * self.team_size)
    
    def __str__(self):
        base = f"{self.name}: ${self.get_salary():,.2f}/year"
        return f"{base} (Senior {self.language} - leads {self.team_size} devs)"

# Test
emp = Employee("Alice", 50000)
mgr = Manager("Bob", 70000, 15000)
dev = Developer("Charlie", 80000, "Python")
senior = SeniorDeveloper("Diana", 100000, "Python", 5)

print("Employee Salaries:")
for person in [emp, mgr, dev, senior]:
    print(f"  {person}")
````

## 3. Multiple Inheritance

````python
# A class can inherit from multiple parents

class Swimmer:
    def swim(self):
        return f"{self.name} is swimming"

class Flyer:
    def fly(self):
        return f"{self.name} is flying"

class Walker:
    def walk(self):
        return f"{self.name} is walking"

# Duck inherits from multiple classes
class Duck(Animal, Swimmer, Flyer, Walker):
    def speak(self):
        return "Quack!"

# Penguin can swim and walk, but not fly
class Penguin(Animal, Swimmer, Walker):
    def speak(self):
        return "Honk!"

# Test
duck = Duck("Donald", 2)
penguin = Penguin("Pingu", 3)

print(f"Duck: {duck.describe()}")
print(f"  {duck.swim()}")
print(f"  {duck.fly()}")
print(f"  {duck.walk()}")
print(f"  Says: {duck.speak()}")

print(f"\nPenguin: {penguin.describe()}")
print(f"  {penguin.swim()}")
print(f"  {penguin.walk()}")
print(f"  Says: {penguin.speak()}")
# penguin.fly()  # Would error - no fly method
````

````python
# Method Resolution Order (MRO)
# Python looks for methods in a specific order

print("Duck MRO:")
for cls in Duck.__mro__:
    print(f"  {cls.__name__}")
````

````python
# Practical multiple inheritance: Mixin classes

class JSONMixin:
    """Add JSON serialization capability"""
    def to_json(self):
        import json
        return json.dumps(self.__dict__, indent=2)

class ComparableMixin:
    """Add comparison capability based on a key"""
    def __lt__(self, other):
        return self._compare_key() < other._compare_key()
    
    def __eq__(self, other):
        return self._compare_key() == other._compare_key()
    
    def _compare_key(self):
        raise NotImplementedError("Subclass must implement _compare_key")

class Product(JSONMixin, ComparableMixin):
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def _compare_key(self):
        return self.price
    
    def __repr__(self):
        return f"Product({self.name}, ${self.price})"

# Test
laptop = Product("Laptop", 999)
mouse = Product("Mouse", 29)
keyboard = Product("Keyboard", 79)

# JSON serialization from mixin
print("JSON output:")
print(laptop.to_json())

# Comparison from mixin
products = [laptop, mouse, keyboard]
print(f"\nSorted by price: {sorted(products)}")
````

## 4. Complete Example: E-Commerce System

````python
# Base classes
class Product:
    """Base product class"""
    
    _id_counter = 0
    
    def __init__(self, name, price):
        Product._id_counter += 1
        self.id = Product._id_counter
        self.name = name
        self.price = price
    
    def get_price(self):
        return self.price
    
    def __str__(self):
        return f"[{self.id}] {self.name}: ${self.get_price():.2f}"

# Specialized product types
class DigitalProduct(Product):
    """Downloadable products"""
    
    def __init__(self, name, price, file_size):
        super().__init__(name, price)
        self.file_size = file_size  # MB
    
    def download(self):
        return f"Downloading {self.name} ({self.file_size} MB)..."
    
    def __str__(self):
        return f"{super().__str__()} (Digital - {self.file_size}MB)"

class PhysicalProduct(Product):
    """Products that need shipping"""
    
    SHIPPING_RATE = 0.05  # $0.05 per gram
    
    def __init__(self, name, price, weight):
        super().__init__(name, price)
        self.weight = weight  # grams
    
    @property
    def shipping_cost(self):
        return self.weight * PhysicalProduct.SHIPPING_RATE
    
    def get_price(self):
        # Include shipping
        return super().get_price() + self.shipping_cost
    
    def __str__(self):
        base_price = super().get_price() - self.shipping_cost
        return f"[{self.id}] {self.name}: ${base_price:.2f} + ${self.shipping_cost:.2f} shipping"

class SubscriptionProduct(DigitalProduct):
    """Recurring subscription"""
    
    def __init__(self, name, monthly_price, file_size):
        super().__init__(name, monthly_price, file_size)
        self.monthly_price = monthly_price
    
    def get_annual_price(self):
        return self.monthly_price * 12 * 0.8  # 20% annual discount
    
    def __str__(self):
        return f"[{self.id}] {self.name}: ${self.price:.2f}/month (Subscription)"

# Demo
print("ðŸ›’ E-COMMERCE PRODUCT CATALOG")
print("=" * 50)

products = [
    DigitalProduct("Python Course", 49.99, 500),
    PhysicalProduct("Python Book", 39.99, 800),
    SubscriptionProduct("Cloud Storage", 9.99, 0),
    PhysicalProduct("Laptop", 999.99, 2500)
]

print("\nðŸ“¦ Products:")
for p in products:
    print(f"  {p}")

print("\nðŸ’µ Total prices (including shipping where applicable):")
for p in products:
    print(f"  {p.name}: ${p.get_price():.2f}")

# Test specific features
course = products[0]
print(f"\nðŸ“¥ {course.download()}")

subscription = products[2]
print(f"\nðŸ“… {subscription.name} annual: ${subscription.get_annual_price():.2f}")
````

## Summary

### Inheritance Syntax:

```python
class Parent:
    pass

class Child(Parent):  # Single inheritance
    pass

class Child(Parent1, Parent2):  # Multiple inheritance
    pass
```

### Key Concepts:

| Concept | Description |
|---------|-------------|
| Inheritance | Child gets all parent attributes/methods |
| Override | Child redefines parent method |
| `super()` | Call parent class methods |
| MRO | Method Resolution Order |

### Best Practices:
1. Use inheritance for "is-a" relationships
2. Always call `super().__init__()` in child
3. Prefer composition over deep inheritance
4. Use mixins for shared behavior
5. Keep inheritance hierarchies shallow
