## 1. What is a Class?

````python
# A class is a blueprint for creating objects
# Think of it like a cookie cutter - the class is the cutter,
# objects are the cookies

# Simplest class
class Dog:
    pass  # Empty class

# Create objects (instances)
dog1 = Dog()
dog2 = Dog()

print(f"dog1: {dog1}")
print(f"dog2: {dog2}")
print(f"Same object? {dog1 is dog2}")
````

````python
# Class with attributes (data)
class Dog:
    # Class attribute - shared by all instances
    species = "Canis familiaris"

dog1 = Dog()
dog2 = Dog()

print(f"dog1.species: {dog1.species}")
print(f"dog2.species: {dog2.species}")
print(f"Dog.species: {Dog.species}")
````

## 2. The __init__ Method (Constructor)

````python
# __init__ is called when creating a new object
# 'self' refers to the instance being created

class Dog:
    species = "Canis familiaris"  # Class attribute
    
    def __init__(self, name, age):
        # Instance attributes - unique to each object
        self.name = name
        self.age = age

# Create instances with different data
buddy = Dog("Buddy", 5)
max_dog = Dog("Max", 3)

print(f"{buddy.name} is {buddy.age} years old")
print(f"{max_dog.name} is {max_dog.age} years old")
````

````python
# Default parameter values
class Dog:
    def __init__(self, name, age=1, breed="Unknown"):
        self.name = name
        self.age = age
        self.breed = breed

# Different ways to create
dog1 = Dog("Buddy")  # Uses defaults
dog2 = Dog("Max", 5)  # Override age
dog3 = Dog("Luna", 3, "Labrador")  # All params

print(f"{dog1.name}: {dog1.age} years, {dog1.breed}")
print(f"{dog2.name}: {dog2.age} years, {dog2.breed}")
print(f"{dog3.name}: {dog3.age} years, {dog3.breed}")
````

## 3. Instance Methods

````python
# Methods are functions inside a class
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        """Make the dog bark"""
        print(f"{self.name} says: Woof!")
    
    def describe(self):
        """Return a description"""
        return f"{self.name} is {self.age} years old"
    
    def birthday(self):
        """Increase age by 1"""
        self.age += 1
        print(f"Happy Birthday {self.name}! Now {self.age} years old.")

buddy = Dog("Buddy", 5)
buddy.bark()
print(buddy.describe())
buddy.birthday()
````

````python
# Methods that return values
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)
    
    def is_square(self):
        return self.width == self.height

rect = Rectangle(10, 5)
print(f"Dimensions: {rect.width} Ã— {rect.height}")
print(f"Area: {rect.area()}")
print(f"Perimeter: {rect.perimeter()}")
print(f"Is square? {rect.is_square()}")
````

## 4. Class vs Instance Attributes

````python
class Dog:
    # Class attribute - shared by ALL instances
    species = "Canis familiaris"
    total_dogs = 0
    
    def __init__(self, name):
        # Instance attribute - unique to each
        self.name = name
        Dog.total_dogs += 1  # Modify class attribute

# Create instances
dog1 = Dog("Buddy")
dog2 = Dog("Max")
dog3 = Dog("Luna")

print(f"Total dogs created: {Dog.total_dogs}")
print(f"Species (class level): {Dog.species}")
print(f"dog1.name: {dog1.name}")
print(f"dog2.name: {dog2.name}")
````

````python
# Be careful: instance shadowing class attribute
class Example:
    value = 100  # Class attribute

obj1 = Example()
obj2 = Example()

print(f"Before: obj1.value={obj1.value}, obj2.value={obj2.value}")

obj1.value = 200  # Creates INSTANCE attribute on obj1

print(f"After: obj1.value={obj1.value}, obj2.value={obj2.value}")
print(f"Class: Example.value={Example.value}")
````

## 5. The __str__ and __repr__ Methods

````python
# Without __str__
class DogBasic:
    def __init__(self, name, age):
        self.name = name
        self.age = age

dog = DogBasic("Buddy", 5)
print(dog)  # Not very helpful!
````

````python
# With __str__ and __repr__
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def __str__(self):
        """Human-readable string (for print)"""
        return f"{self.name}, {self.age} years old"
    
    def __repr__(self):
        """Developer-readable (for debugging)"""
        return f"Dog('{self.name}', {self.age})"

dog = Dog("Buddy", 5)

print(f"str: {str(dog)}")
print(f"repr: {repr(dog)}")
print(f"print: {dog}")
````

## 6. Complete Example: Bank Account

````python
class BankAccount:
    """
    A simple bank account class.
    
    Demonstrates:
    - Constructor
    - Instance methods
    - Data validation
    - Transaction history
    """
    
    # Class attribute
    bank_name = "Python Bank"
    
    def __init__(self, owner, balance=0):
        """Initialize account with owner and optional balance"""
        self.owner = owner
        self.balance = balance
        self.transactions = []
        self._log_transaction("Account opened", balance)
    
    def deposit(self, amount):
        """Deposit money into account"""
        if amount <= 0:
            print("âŒ Deposit amount must be positive")
            return False
        
        self.balance += amount
        self._log_transaction("Deposit", amount)
        print(f"âœ… Deposited ${amount:.2f}")
        return True
    
    def withdraw(self, amount):
        """Withdraw money from account"""
        if amount <= 0:
            print("âŒ Withdrawal amount must be positive")
            return False
        
        if amount > self.balance:
            print("âŒ Insufficient funds")
            return False
        
        self.balance -= amount
        self._log_transaction("Withdrawal", -amount)
        print(f"âœ… Withdrew ${amount:.2f}")
        return True
    
    def get_balance(self):
        """Return current balance"""
        return self.balance
    
    def _log_transaction(self, description, amount):
        """Log transaction (internal method)"""
        from datetime import datetime
        self.transactions.append({
            "date": datetime.now().strftime("%Y-%m-%d %H:%M"),
            "description": description,
            "amount": amount,
            "balance": self.balance
        })
    
    def show_transactions(self):
        """Display transaction history"""
        print(f"\nðŸ“‹ Transaction History for {self.owner}:")
        print("-" * 50)
        for t in self.transactions:
            sign = "+" if t["amount"] >= 0 else ""
            print(f"{t['date']} | {t['description']:<15} | "
                  f"{sign}${t['amount']:.2f} | Balance: ${t['balance']:.2f}")
    
    def __str__(self):
        return f"{self.owner}'s account: ${self.balance:.2f}"
    
    def __repr__(self):
        return f"BankAccount('{self.owner}', {self.balance})"

# Demo
print("=" * 50)
print(f"Welcome to {BankAccount.bank_name}")
print("=" * 50)

# Create account
account = BankAccount("Alice", 100)
print(f"\n{account}")

# Operations
account.deposit(500)
account.withdraw(200)
account.deposit(50)
account.withdraw(1000)  # Should fail

# Show history
account.show_transactions()

print(f"\nFinal balance: ${account.get_balance():.2f}")
````

## Summary

### OOP Terminology:

| Term | Description |
|------|-------------|
| Class | Blueprint/template |
| Object | Instance of a class |
| Attribute | Data stored in object |
| Method | Function inside class |
| Constructor | `__init__` method |
| self | Reference to current instance |

### Class Structure:
```python
class ClassName:
    class_attribute = value  # Shared
    
    def __init__(self, params):
        self.instance_attr = value  # Per instance
    
    def method(self):
        # Uses self.attributes
        pass
```

### Special Methods:

| Method | Purpose |
|--------|----------|
| `__init__` | Initialize new object |
| `__str__` | String for users |
| `__repr__` | String for developers |
