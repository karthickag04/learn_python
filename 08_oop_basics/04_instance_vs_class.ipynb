## 1. Instance Variables vs Class Variables

```python
# Instance variables: unique to each object
# Class variables: shared by all objects

class Student:
    # Class variable - shared by all instances
    school = "Python Academy"
    student_count = 0
    
    def __init__(self, name, grade):
        # Instance variables - unique to each student
        self.name = name
        self.grade = grade
        
        # Update class variable
        Student.student_count += 1

# Create students
s1 = Student("Alice", "A")
s2 = Student("Bob", "B")
s3 = Student("Charlie", "A")

# Instance variables are different
print(f"Student 1: {s1.name}, Grade: {s1.grade}")
print(f"Student 2: {s2.name}, Grade: {s2.grade}")

# Class variable is same for all
print(f"\nSchool (via s1): {s1.school}")
print(f"School (via s2): {s2.school}")
print(f"School (via class): {Student.school}")

# Count is shared
print(f"\nTotal students: {Student.student_count}")
```

```python
# Important: Modifying class variables

class Example:
    shared_list = []  # Class variable
    shared_value = 0
    
    def __init__(self, name):
        self.name = name

e1 = Example("first")
e2 = Example("second")

# Modifying mutable class variable affects ALL instances
e1.shared_list.append("added via e1")
print(f"e1.shared_list: {e1.shared_list}")
print(f"e2.shared_list: {e2.shared_list}")  # Same!

# But assignment creates instance variable (shadow)
e1.shared_value = 100  # Creates new instance variable!
print(f"\ne1.shared_value: {e1.shared_value}")
print(f"e2.shared_value: {e2.shared_value}")  # Still 0
print(f"Example.shared_value: {Example.shared_value}")  # Still 0

# Check what attributes each has
print(f"\ne1 has own shared_value: {'shared_value' in e1.__dict__}")
print(f"e2 has own shared_value: {'shared_value' in e2.__dict__}")
```

## 2. Method Types

```python
class Calculator:
    # Class variable
    history = []
    
    def __init__(self, name="Basic"):
        self.name = name
        self.results = []  # Instance variable
    
    # Instance method - has access to self
    def add(self, a, b):
        result = a + b
        self.results.append(f"{a} + {b} = {result}")
        Calculator.history.append(result)
        return result
    
    # Class method - has access to cls (the class)
    @classmethod
    def get_history(cls):
        return cls.history
    
    @classmethod
    def clear_history(cls):
        cls.history.clear()
        print("Global history cleared")
    
    # Static method - no access to self or cls
    @staticmethod
    def is_even(n):
        return n % 2 == 0
    
    @staticmethod
    def validate_input(value):
        return isinstance(value, (int, float))

# Create calculators
calc1 = Calculator("Scientific")
calc2 = Calculator("Simple")

# Instance methods
calc1.add(5, 3)
calc1.add(10, 20)
calc2.add(1, 1)

print(f"calc1 results: {calc1.results}")
print(f"calc2 results: {calc2.results}")

# Class method - same result from any instance or class
print(f"\nHistory (via class): {Calculator.get_history()}")
print(f"History (via calc1): {calc1.get_history()}")

# Static method - utility function
print(f"\nIs 4 even? {Calculator.is_even(4)}")
print(f"Is 7 even? {calc1.is_even(7)}")
```

```python
# Class methods as alternative constructors

class Date:
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day
    
    @classmethod
    def from_string(cls, date_string):
        """Create from 'YYYY-MM-DD' string"""
        parts = date_string.split('-')
        return cls(int(parts[0]), int(parts[1]), int(parts[2]))
    
    @classmethod
    def today(cls):
        """Create with today's date"""
        from datetime import date
        today = date.today()
        return cls(today.year, today.month, today.day)
    
    @classmethod
    def from_timestamp(cls, timestamp):
        """Create from Unix timestamp"""
        from datetime import datetime
        dt = datetime.fromtimestamp(timestamp)
        return cls(dt.year, dt.month, dt.day)
    
    def __str__(self):
        return f"{self.year}-{self.month:02d}-{self.day:02d}"

# Different ways to create Date
d1 = Date(2024, 12, 25)
d2 = Date.from_string("2024-01-15")
d3 = Date.today()

print(f"Direct: {d1}")
print(f"From string: {d2}")
print(f"Today: {d3}")
```

## 3. When to Use Each

```python
class Employee:
    """
    Demonstrates when to use each type.
    """
    
    # Class variables: data shared by all
    company = "TechCorp"
    tax_rate = 0.22
    _employee_count = 0
    
    def __init__(self, name, salary):
        # Instance variables: unique to each employee
        self.name = name
        self.salary = salary
        self.id = Employee._generate_id()
        Employee._employee_count += 1
    
    # Instance method: operates on specific employee
    def get_net_salary(self):
        """Calculate net salary after tax"""
        return self.salary * (1 - Employee.tax_rate)
    
    def give_raise(self, percent):
        """Give this employee a raise"""
        self.salary *= (1 + percent / 100)
    
    def __str__(self):
        return f"[{self.id}] {self.name} - ${self.salary:.2f}"
    
    # Class method: operates on class data or creates instances
    @classmethod
    def set_tax_rate(cls, rate):
        """Update tax rate for all employees"""
        cls.tax_rate = rate
    
    @classmethod
    def get_headcount(cls):
        """Get total employee count"""
        return cls._employee_count
    
    @classmethod
    def from_csv(cls, csv_line):
        """Create employee from CSV line"""
        name, salary = csv_line.strip().split(',')
        return cls(name, float(salary))
    
    # Static method: utility function, doesn't need self or cls
    @staticmethod
    def _generate_id():
        """Generate unique ID"""
        import random
        return f"EMP{random.randint(1000, 9999)}"
    
    @staticmethod
    def is_valid_salary(salary):
        """Validate salary value"""
        return isinstance(salary, (int, float)) and salary > 0
    
    @staticmethod
    def format_currency(amount):
        """Format as currency"""
        return f"${amount:,.2f}"

# Demo
print("=" * 50)
print("ðŸ‘¥ EMPLOYEE MANAGEMENT DEMO")
print("=" * 50)

# Create employees
e1 = Employee("Alice", 75000)
e2 = Employee("Bob", 65000)
e3 = Employee.from_csv("Charlie,55000")

print(f"\nðŸ“‹ Employees:")
for emp in [e1, e2, e3]:
    print(f"  {emp}")

# Instance method
print(f"\nðŸ’° Net salaries (after {Employee.tax_rate*100}% tax):")
print(f"  Alice: {Employee.format_currency(e1.get_net_salary())}")

# Class method affects all
Employee.set_tax_rate(0.25)
print(f"\nðŸ“Š After tax increase to 25%:")
print(f"  Alice: {Employee.format_currency(e1.get_net_salary())}")

# Class method
print(f"\nðŸ‘¥ Total headcount: {Employee.get_headcount()}")

# Static method
print(f"\nâœ… Is 50000 valid salary? {Employee.is_valid_salary(50000)}")
print(f"âŒ Is -1000 valid salary? {Employee.is_valid_salary(-1000)}")
```

## 4. Complete Example: Product Inventory System

```python
class Product:
    """
    Complete inventory system demonstrating all variable and method types.
    """
    
    # Class variables
    store_name = "PyMart"
    default_tax = 0.08
    _all_products = {}  # ID -> Product
    _next_id = 1
    
    def __init__(self, name, price, quantity):
        # Instance variables
        self.product_id = Product._next_id
        self.name = name
        self._price = price
        self._quantity = quantity
        
        # Register product
        Product._all_products[self.product_id] = self
        Product._next_id += 1
    
    # Properties (instance level)
    @property
    def price(self):
        return self._price
    
    @price.setter
    def price(self, value):
        if value < 0:
            raise ValueError("Price cannot be negative")
        self._price = value
    
    @property
    def quantity(self):
        return self._quantity
    
    @property
    def value(self):
        """Total value of stock"""
        return self._price * self._quantity
    
    @property
    def in_stock(self):
        return self._quantity > 0
    
    # Instance methods
    def sell(self, qty):
        """Sell items"""
        if qty > self._quantity:
            raise ValueError(f"Only {self._quantity} available")
        self._quantity -= qty
        total = qty * self._price * (1 + Product.default_tax)
        return total
    
    def restock(self, qty):
        """Add stock"""
        if qty < 0:
            raise ValueError("Cannot add negative quantity")
        self._quantity += qty
    
    def __str__(self):
        status = "âœ“" if self.in_stock else "âœ—"
        return f"[{self.product_id}] {self.name}: ${self._price:.2f} x {self._quantity} {status}"
    
    # Class methods
    @classmethod
    def find_by_id(cls, product_id):
        """Find product by ID"""
        return cls._all_products.get(product_id)
    
    @classmethod
    def get_all_products(cls):
        """Get all products"""
        return list(cls._all_products.values())
    
    @classmethod
    def total_inventory_value(cls):
        """Calculate total inventory value"""
        return sum(p.value for p in cls._all_products.values())
    
    @classmethod
    def low_stock_report(cls, threshold=5):
        """Find products with low stock"""
        return [p for p in cls._all_products.values() if p.quantity <= threshold]
    
    @classmethod
    def from_dict(cls, data):
        """Create from dictionary"""
        return cls(data['name'], data['price'], data['quantity'])
    
    # Static methods
    @staticmethod
    def calculate_discount(price, percent):
        """Calculate discounted price"""
        return price * (1 - percent / 100)
    
    @staticmethod
    def format_price(amount):
        """Format price"""
        return f"${amount:,.2f}"

# Demo
print("=" * 60)
print(f"ðŸª {Product.store_name} INVENTORY SYSTEM")
print("=" * 60)

# Create products
products = [
    Product("Laptop", 999.99, 15),
    Product("Mouse", 29.99, 50),
    Product("Keyboard", 79.99, 3),
    Product("Monitor", 349.99, 8),
    Product("Webcam", 89.99, 0)
]

# List all products
print("\nðŸ“¦ PRODUCTS:")
for p in Product.get_all_products():
    print(f"  {p}")

# Total value
print(f"\nðŸ’° Total inventory value: {Product.format_price(Product.total_inventory_value())}")

# Low stock
print("\nâš ï¸ LOW STOCK ALERT (â‰¤5 items):")
for p in Product.low_stock_report():
    print(f"  {p.name}: only {p.quantity} left!")

# Make a sale
laptop = Product.find_by_id(1)
sale_total = laptop.sell(2)
print(f"\nðŸ›’ Sold 2 Laptops: {Product.format_price(sale_total)} (incl. tax)")
print(f"   Remaining: {laptop.quantity}")

# Calculate discount
print(f"\nðŸ·ï¸ 20% off Monitor: {Product.format_price(Product.calculate_discount(349.99, 20))}")
```

## Summary

### Variable Types:

| Type | Definition | Scope | Example |
|------|------------|-------|----------|
| Instance | `self.var` in method | Unique per object | `self.name` |
| Class | In class body | Shared by all | `student_count` |

### Method Types:

| Type | Decorator | First Param | Use Case |
|------|-----------|-------------|----------|
| Instance | None | `self` | Work with instance data |
| Class | `@classmethod` | `cls` | Work with class data, factories |
| Static | `@staticmethod` | None | Utility functions |

### When to Use:

| Scenario | Use |
|----------|-----|
| Data unique to each object | Instance variable |
| Data shared across all objects | Class variable |
| Method needs instance data | Instance method |
| Method needs class data, not instance | Class method |
| Alternative constructor | Class method |
| Pure utility, no state needed | Static method |
