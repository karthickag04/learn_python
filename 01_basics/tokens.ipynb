# Python Tokens: A Beginner's Guide

Python code is made up of tokens. Tokens are the smallest units in a Python program that have meaning. Understanding tokens is essential for reading and writing Python code. Here’s a topic-wise breakdown with short explanations.

## 1. Keywords

**Definition:** Reserved words that have special meaning in Python and cannot be used as identifiers (variable names, function names, etc).

**Examples:**
- `if`, `else`, `elif`, `for`, `while`, `break`, `continue`, `def`, `return`, `import`, `from`, `as`, `class`, `try`, `except`, `finally`, `with`, `pass`, `raise`, `yield`, `global`, `nonlocal`, `assert`, `lambda`, `del`, `in`, `is`, `not`, `or`, `and`, `True`, `False`, `None`

**Usage:**
```python
if x > 0:
    print("Positive")
else:
    print("Not positive")
```

## 2. Identifiers

**Definition:** Names given to variables, functions, classes, etc. Identifiers must start with a letter (A-Z or a-z) or underscore (_), followed by letters, digits (0-9), or underscores.

**Examples:**
- `age`, `total_sum`, `_private_var`, `MyClass`

**Usage:**
```python
name = "Alice"
score = 95
```


## 3. Literals

**Definition:** Fixed values assigned to variables. They represent constant values in code.

**Types of Literals:**
- **String Literals:** Text values, e.g., `'hello'`, `"world"`
- **Numeric Literals:** Numbers, e.g., `10`, `3.14`, `0b101` (binary), `0x1A` (hex)
- **Boolean Literals:** `True`, `False`
- **None Literal:** `None` (represents no value)

**Usage:**
```python
name = "Bob"      # String literal
age = 25          # Integer literal
pi = 3.14         # Float literal
is_valid = True   # Boolean literal
nothing = None    # None literal
```

## 4. Operators

**Definition:** Symbols that perform operations on variables and values.

**Types of Operators:**
- **Arithmetic Operators:** `+`, `-`, `*`, `/`, `//`, `%`, `**`
- **Assignment Operators:** `=`, `+=`, `-=`, `*=`, `/=`, etc.
- **Comparison Operators:** `==`, `!=`, `>`, `<`, `>=`, `<=`
- **Logical Operators:** `and`, `or`, `not`
- **Bitwise Operators:** `&`, `|`, `^`, `~`, `<<`, `>>`
- **Membership Operators:** `in`, `not in`
- **Identity Operators:** `is`, `is not`

**Usage:**
```python
a = 10
b = 3
print(a + b)      # Arithmetic
print(a > b)      # Comparison
print(a == 10 and b < 5)  # Logical
```

## 5. Delimiters (Separators)

**Definition:** Characters that separate statements, expressions, or items in Python code.

**Examples:**
- Parentheses: `(`, `)`
- Brackets: `[`, `]`
- Braces: `{`, `}`
- Comma: `,`
- Colon: `:`
- Dot: `.`
- Semicolon: `;`
- At symbol: `@`
- Assignment: `=`

**Usage:**
```python
list1 = [1, 2, 3]
def greet(name):
    print("Hello,", name)
```

## 6. Comments

**Definition:** Notes in the code that are ignored by Python. Used to explain code and make it more readable.

**Types:**
- **Single-line comment:** Starts with `#`
- **Multi-line comment:** Triple quotes (`'''` or `"""`), usually for docstrings

**Usage:**
```python
# This is a single-line comment
'''
This is a multi-line comment or docstring
'''
```

## Summary Table

| Token Type   | Description                                      | Examples |
|--------------|--------------------------------------------------|----------|
| Keywords     | Reserved words with special meaning              | `if`, `for`, `def` |
| Identifiers  | Names for variables, functions, etc.             | `x`, `my_var` |
| Literals     | Fixed values                                     | `10`, `"hello"`, `True` |
| Operators    | Symbols for operations                           | `+`, `-`, `*`, `==` |
| Delimiters   | Characters that separate code elements           | `,`, `:`, `()`, `[]` |
| Comments     | Notes for code explanation, ignored by Python    | `# comment` |
