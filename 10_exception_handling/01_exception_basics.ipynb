## 1. What are Exceptions?

````python
# Exceptions are errors that occur during program execution
# Without handling, they crash your program

# Common exception examples:
print("Common exception types:")

# ZeroDivisionError
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"  ZeroDivisionError: {e}")

# TypeError
try:
    result = "hello" + 5
except TypeError as e:
    print(f"  TypeError: {e}")

# ValueError
try:
    number = int("abc")
except ValueError as e:
    print(f"  ValueError: {e}")

# KeyError
try:
    d = {"a": 1}
    value = d["b"]
except KeyError as e:
    print(f"  KeyError: {e}")

# IndexError
try:
    lst = [1, 2, 3]
    item = lst[10]
except IndexError as e:
    print(f"  IndexError: {e}")

# AttributeError
try:
    x = None
    x.something()
except AttributeError as e:
    print(f"  AttributeError: {e}")
````

## 2. Basic try/except

````python
# Basic syntax
def divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Cannot divide by zero!"

print(f"10 / 2 = {divide(10, 2)}")
print(f"10 / 0 = {divide(10, 0)}")
````

````python
# Capturing the exception object
def safe_convert(value):
    try:
        return int(value)
    except ValueError as e:
        print(f"Conversion error: {e}")
        return None

print(safe_convert("42"))      # Works
print(safe_convert("abc"))     # Error handled
print(safe_convert("3.14"))    # Error handled
````

````python
# Real-world example: User input validation
def get_age():
    while True:
        user_input = input("Enter your age: ")
        try:
            age = int(user_input)
            if age < 0 or age > 150:
                print("Please enter a valid age (0-150)")
                continue
            return age
        except ValueError:
            print("Invalid input. Please enter a number.")

# Simulated test (comment out for interactive use)
print("Age validation example (simulated):")
test_inputs = ["abc", "-5", "25"]
for inp in test_inputs:
    print(f"  Input: '{inp}'")
    try:
        age = int(inp)
        if 0 <= age <= 150:
            print(f"  Valid age: {age}")
        else:
            print(f"  Age out of range")
    except ValueError:
        print(f"  Not a number")
````

## 3. Multiple Exception Types

````python
# Handle different exceptions differently
def process_data(data, index):
    try:
        item = data[index]
        result = 100 / item
        return result
    except IndexError:
        return f"Error: Index {index} out of range"
    except ZeroDivisionError:
        return "Error: Division by zero"
    except TypeError:
        return "Error: Invalid data type"

data = [10, 0, 5, "abc"]

print("Processing data:")
print(f"  Index 0: {process_data(data, 0)}")
print(f"  Index 1: {process_data(data, 1)}")
print(f"  Index 2: {process_data(data, 2)}")
print(f"  Index 3: {process_data(data, 3)}")
print(f"  Index 10: {process_data(data, 10)}")
````

````python
# Handle multiple exceptions with same handler
def parse_number(value):
    try:
        if '.' in str(value):
            return float(value)
        return int(value)
    except (ValueError, TypeError) as e:
        print(f"Could not parse '{value}': {e}")
        return None

print(parse_number("42"))
print(parse_number("3.14"))
print(parse_number("hello"))
print(parse_number(None))
````

````python
# Catch-all exception (use sparingly)
def risky_operation(data):
    try:
        result = data['value'] * 2
        return result
    except Exception as e:
        # Logs the error but doesn't expose internal details
        print(f"An error occurred: {type(e).__name__}")
        return None

print(risky_operation({'value': 10}))  # Works
print(risky_operation({}))              # KeyError
print(risky_operation(None))            # TypeError
````

## 4. else and finally Clauses

````python
# else: runs only if no exception occurred
# finally: always runs

def divide_and_log(a, b):
    print(f"\nDividing {a} by {b}:")
    try:
        result = a / b
    except ZeroDivisionError:
        print("  ERROR: Division by zero")
        result = None
    else:
        print(f"  SUCCESS: Result = {result}")
    finally:
        print("  CLEANUP: Operation complete")
    
    return result

divide_and_log(10, 2)
divide_and_log(10, 0)
````

````python
# finally is great for cleanup
def process_file_simulation(filename):
    file_handle = None
    try:
        print(f"Opening {filename}...")
        file_handle = f"FILE:{filename}"  # Simulated file handle
        
        if "error" in filename:
            raise IOError("Simulated file error")
        
        print(f"Processing {filename}...")
        return "Success"
    
    except IOError as e:
        print(f"Error: {e}")
        return "Failed"
    
    finally:
        if file_handle:
            print(f"Closing {filename}...")  # Always runs!

print("Test 1:")
process_file_simulation("data.txt")

print("\nTest 2:")
process_file_simulation("error_file.txt")
````

## 5. Complete Example: Safe Calculator

````python
class SafeCalculator:
    """
    Calculator with comprehensive error handling.
    """
    
    def __init__(self):
        self.history = []
    
    def _validate_numbers(self, *args):
        """Validate that all arguments are numbers"""
        for arg in args:
            if not isinstance(arg, (int, float)):
                raise TypeError(f"Expected number, got {type(arg).__name__}")
    
    def _log(self, operation, result):
        """Log operation to history"""
        self.history.append(f"{operation} = {result}")
    
    def add(self, a, b):
        try:
            self._validate_numbers(a, b)
            result = a + b
        except TypeError as e:
            return f"Error: {e}"
        else:
            self._log(f"{a} + {b}", result)
            return result
    
    def subtract(self, a, b):
        try:
            self._validate_numbers(a, b)
            result = a - b
        except TypeError as e:
            return f"Error: {e}"
        else:
            self._log(f"{a} - {b}", result)
            return result
    
    def multiply(self, a, b):
        try:
            self._validate_numbers(a, b)
            result = a * b
        except TypeError as e:
            return f"Error: {e}"
        else:
            self._log(f"{a} Ã— {b}", result)
            return result
    
    def divide(self, a, b):
        try:
            self._validate_numbers(a, b)
            if b == 0:
                raise ZeroDivisionError("Cannot divide by zero")
            result = a / b
        except (TypeError, ZeroDivisionError) as e:
            return f"Error: {e}"
        else:
            self._log(f"{a} Ã· {b}", result)
            return result
    
    def power(self, base, exp):
        try:
            self._validate_numbers(base, exp)
            result = base ** exp
        except TypeError as e:
            return f"Error: {e}"
        except OverflowError:
            return "Error: Result too large"
        else:
            self._log(f"{base}^{exp}", result)
            return result
    
    def sqrt(self, n):
        import math
        try:
            self._validate_numbers(n)
            if n < 0:
                raise ValueError("Cannot take square root of negative number")
            result = math.sqrt(n)
        except (TypeError, ValueError) as e:
            return f"Error: {e}"
        else:
            self._log(f"âˆš{n}", result)
            return result
    
    def show_history(self):
        print("\nðŸ“ Calculation History:")
        if not self.history:
            print("  (empty)")
        else:
            for entry in self.history[-10:]:
                print(f"  {entry}")

# Demo
print("ðŸ”¢ SAFE CALCULATOR DEMO")
print("=" * 40)

calc = SafeCalculator()

# Valid operations
print(f"10 + 5 = {calc.add(10, 5)}")
print(f"10 - 5 = {calc.subtract(10, 5)}")
print(f"10 Ã— 5 = {calc.multiply(10, 5)}")
print(f"10 Ã· 5 = {calc.divide(10, 5)}")
print(f"2^10 = {calc.power(2, 10)}")
print(f"âˆš16 = {calc.sqrt(16)}")

# Error cases
print("\n--- Error Cases ---")
print(f"10 Ã· 0 = {calc.divide(10, 0)}")
print(f"10 + 'five' = {calc.add(10, 'five')}")
print(f"âˆš-4 = {calc.sqrt(-4)}")

calc.show_history()
````

## Summary

### Exception Handling Syntax:

```python
try:
    # Code that might raise exception
except ExceptionType as e:
    # Handle specific exception
except (Type1, Type2):
    # Handle multiple types
except Exception:
    # Catch all (use carefully)
else:
    # Runs if no exception
finally:
    # Always runs (cleanup)
```

### Common Exceptions:

| Exception | Cause |
|-----------|-------|
| `ValueError` | Wrong value type |
| `TypeError` | Wrong data type |
| `KeyError` | Missing dictionary key |
| `IndexError` | List index out of range |
| `ZeroDivisionError` | Division by zero |
| `FileNotFoundError` | File doesn't exist |
| `AttributeError` | Missing attribute |

### Best Practices:
1. Catch specific exceptions, not generic `Exception`
2. Use `else` for code that should run on success
3. Use `finally` for cleanup (closing files, connections)
4. Don't silence exceptions without handling them
5. Provide useful error messages
