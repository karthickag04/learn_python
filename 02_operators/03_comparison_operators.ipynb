## 1. Equal (==)

```python
# Equal (==) - Checks if two values are the same
print("=== EQUAL (==) ===")

# Numbers
print(f"5 == 5: {5 == 5}")    # True
print(f"5 == 3: {5 == 3}")    # False

# Strings
print(f"'hello' == 'hello': {'hello' == 'hello'}")  # True
print(f"'Hello' == 'hello': {'Hello' == 'hello'}")  # False (case sensitive)

# Variables
x = 10
y = 10
z = 20
print(f"\nx = {x}, y = {y}, z = {z}")
print(f"x == y: {x == y}")  # True
print(f"x == z: {x == z}")  # False
```

```python
# Real-world: Password Check
stored_password = "secret123"
entered_password = "secret123"

if entered_password == stored_password:
    print("âœ… Password correct! Access granted.")
else:
    print("âŒ Wrong password! Access denied.")

# Try wrong password
entered_password = "wrongpass"
if entered_password == stored_password:
    print("âœ… Password correct! Access granted.")
else:
    print("âŒ Wrong password! Access denied.")
```

## 2. Not Equal (!=)

```python
# Not Equal (!=) - Checks if two values are different
print("=== NOT EQUAL (!=) ===")

print(f"5 != 3: {5 != 3}")    # True (they are different)
print(f"5 != 5: {5 != 5}")    # False (they are same)

# Strings
print(f"'cat' != 'dog': {'cat' != 'dog'}")  # True
print(f"'cat' != 'cat': {'cat' != 'cat'}")  # False
```

```python
# Real-world: Input Validation
user_input = "quit"

# Keep asking while input is not 'quit'
commands = ["help", "status", "run", "quit"]

for cmd in commands:
    if cmd != "quit":
        print(f"Processing command: {cmd}")
    else:
        print("Exiting program...")
```

## 3. Greater Than (>)

```python
# Greater Than (>) - Checks if left is greater than right
print("=== GREATER THAN (>) ===")

print(f"10 > 5: {10 > 5}")    # True
print(f"5 > 10: {5 > 10}")    # False
print(f"5 > 5: {5 > 5}")      # False (must be greater, not equal)
```

```python
# Real-world: Age Verification
user_age = 21
drinking_age = 21

print(f"User Age: {user_age}")

if user_age > 18:
    print("âœ… You can vote!")

if user_age > 16:
    print("âœ… You can drive!")

if user_age > 25:
    print("âœ… You can rent a car without extra fees!")
else:
    print("âŒ Extra fees apply for car rental under 25.")
```

## 4. Less Than (<)

```python
# Less Than (<) - Checks if left is less than right
print("=== LESS THAN (<) ===")

print(f"5 < 10: {5 < 10}")    # True
print(f"10 < 5: {10 < 5}")    # False
print(f"5 < 5: {5 < 5}")      # False (must be less, not equal)
```

```python
# Real-world: Inventory Warning
stock_quantity = 8
low_stock_threshold = 10
critical_threshold = 5

print(f"Current Stock: {stock_quantity} units")

if stock_quantity < critical_threshold:
    print("ğŸ”´ CRITICAL: Stock critically low! Order immediately!")
elif stock_quantity < low_stock_threshold:
    print("ğŸŸ¡ WARNING: Low stock! Consider reordering.")
else:
    print("ğŸŸ¢ Stock level is healthy.")
```

## 5. Greater Than or Equal (>=)

```python
# Greater Than or Equal (>=)
print("=== GREATER THAN OR EQUAL (>=) ===")

print(f"10 >= 5: {10 >= 5}")   # True (10 is greater)
print(f"5 >= 5: {5 >= 5}")     # True (5 equals 5)
print(f"3 >= 5: {3 >= 5}")     # False
```

```python
# Real-world: Discount Eligibility
order_total = 75.00
free_shipping_threshold = 50.00
vip_threshold = 100.00

print(f"Order Total: ${order_total:.2f}")

if order_total >= vip_threshold:
    print("ğŸŒŸ VIP discount applied! 20% off + Free shipping")
elif order_total >= free_shipping_threshold:
    print("ğŸ“¦ Free shipping unlocked!")
else:
    remaining = free_shipping_threshold - order_total
    print(f"ğŸ’¡ Add ${remaining:.2f} more for free shipping!")
```

## 6. Less Than or Equal (<=)

```python
# Less Than or Equal (<=)
print("=== LESS THAN OR EQUAL (<=) ===")

print(f"5 <= 10: {5 <= 10}")   # True (5 is less)
print(f"5 <= 5: {5 <= 5}")     # True (5 equals 5)
print(f"10 <= 5: {10 <= 5}")   # False
```

```python
# Real-world: Capacity Check
current_passengers = 48
max_capacity = 50

print(f"Current Passengers: {current_passengers}")
print(f"Maximum Capacity: {max_capacity}")

if current_passengers <= max_capacity:
    available = max_capacity - current_passengers
    print(f"âœ… Boarding allowed! {available} seats available.")
else:
    print("âŒ Overcapacity! Cannot board more passengers.")
```

## 7. Chained Comparisons

```python
# Python allows chaining comparisons
print("=== CHAINED COMPARISONS ===")

x = 5

# Check if x is between 1 and 10
print(f"x = {x}")
print(f"1 < x < 10: {1 < x < 10}")      # True
print(f"1 <= x <= 10: {1 <= x <= 10}")  # True
print(f"5 < x < 10: {5 < x < 10}")      # False (x is not > 5)
print(f"5 <= x < 10: {5 <= x < 10}")    # True
```

```python
# Real-world: Grade Classification
score = 78
print(f"Student Score: {score}")

if 90 <= score <= 100:
    grade = "A"
elif 80 <= score < 90:
    grade = "B"
elif 70 <= score < 80:
    grade = "C"
elif 60 <= score < 70:
    grade = "D"
else:
    grade = "F"

print(f"Grade: {grade}")
```

## 8. Comparing Different Types

```python
# Comparing strings (alphabetical order)
print("=== STRING COMPARISONS ===")

print(f"'apple' < 'banana': {'apple' < 'banana'}")  # True (a comes before b)
print(f"'Apple' < 'apple': {'Apple' < 'apple'}")    # True (uppercase < lowercase)
print(f"'abc' < 'abd': {'abc' < 'abd'}")            # True

# Useful for sorting names
names = ["Charlie", "Alice", "Bob"]
sorted_names = sorted(names)
print(f"\nOriginal: {names}")
print(f"Sorted: {sorted_names}")
```

```python
# Comparing lists
print("=== LIST COMPARISONS ===")

list1 = [1, 2, 3]
list2 = [1, 2, 3]
list3 = [1, 2, 4]

print(f"list1: {list1}")
print(f"list2: {list2}")
print(f"list3: {list3}")
print(f"\nlist1 == list2: {list1 == list2}")  # True
print(f"list1 == list3: {list1 == list3}")    # False
print(f"list1 < list3: {list1 < list3}")      # True (compares element by element)
```

## 9. Complete Example: Student Admission System

```python
# Real-world: Complete Student Admission System

def check_admission(name, age, score, income):
    print(f"\n{'='*40}")
    print(f"Checking Admission for: {name}")
    print(f"Age: {age} | Score: {score} | Family Income: ${income:,}")
    print("-" * 40)
    
    # Age requirement
    if age < 17:
        print("âŒ Age requirement not met (minimum 17)")
        return False
    
    # Score requirement
    if score < 60:
        print("âŒ Score requirement not met (minimum 60)")
        return False
    
    print("âœ… Basic requirements met!")
    
    # Scholarship eligibility
    if score >= 90 and income <= 50000:
        print("ğŸŒŸ FULL SCHOLARSHIP GRANTED!")
    elif score >= 80:
        print("ğŸ“š Partial scholarship available (50%)")
    elif score >= 70:
        print("ğŸ“– Merit-based discount (20%)")
    else:
        print("ğŸ’° Standard tuition applies")
    
    # Housing priority
    if age >= 18 and score >= 75:
        print("ğŸ  Priority housing available!")
    
    return True

# Test different students
students = [
    ("Alice", 18, 92, 45000),
    ("Bob", 17, 78, 80000),
    ("Charlie", 16, 85, 60000),
    ("Diana", 19, 55, 40000),
]

for student in students:
    check_admission(*student)
```

## Summary

| Operator | Name | Example | Use Case |
|----------|------|---------|----------|
| `==` | Equal | `x == 5` | Check equality |
| `!=` | Not Equal | `x != 5` | Check difference |
| `>` | Greater | `x > 5` | Check if larger |
| `<` | Less | `x < 5` | Check if smaller |
| `>=` | Greater/Equal | `x >= 5` | Thresholds |
| `<=` | Less/Equal | `x <= 5` | Limits |

### Key Points:
1. Comparison operators always return `True` or `False`
2. `==` checks equality, `=` assigns values (don't confuse!)
3. Python allows chaining: `1 < x < 10`
4. Strings are compared alphabetically
5. Used extensively in if statements and loops
