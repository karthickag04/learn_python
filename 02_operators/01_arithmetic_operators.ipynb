## 1. Addition (+)

```python
# ADDITION (+) - Adds two values
a = 10
b = 5

result = a + b
print(f"{a} + {b} = {result}")

# Real-world: Shopping cart total
item1_price = 299.99
item2_price = 149.50
item3_price = 49.99

total = item1_price + item2_price + item3_price
print(f"\nShopping Cart Total: ${total:.2f}")
```

```python
# String concatenation with +
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(f"Full Name: {full_name}")

# List concatenation
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined = list1 + list2
print(f"Combined List: {combined}")
```

## 2. Subtraction (-)

```python
# SUBTRACTION (-) - Subtracts second value from first
a = 20
b = 8

result = a - b
print(f"{a} - {b} = {result}")

# Real-world: Calculate remaining balance
account_balance = 1000.00
withdrawal = 250.00
remaining = account_balance - withdrawal
print(f"\nAccount Balance: ${account_balance}")
print(f"Withdrawal: ${withdrawal}")
print(f"Remaining: ${remaining}")
```

```python
# Real-world: Calculate discount
original_price = 499.99
discount_amount = 75.00
final_price = original_price - discount_amount

print(f"Original Price: ${original_price}")
print(f"Discount: -${discount_amount}")
print(f"Final Price: ${final_price}")
```

## 3. Multiplication (*)

```python
# MULTIPLICATION (*) - Multiplies two values
a = 7
b = 6

result = a * b
print(f"{a} * {b} = {result}")

# Real-world: Calculate total cost for multiple items
item_price = 29.99
quantity = 5
total_cost = item_price * quantity
print(f"\nItem Price: ${item_price}")
print(f"Quantity: {quantity}")
print(f"Total Cost: ${total_cost:.2f}")
```

```python
# String repetition with *
line = "-" * 40
print(line)
print("   WELCOME TO PYTHON")
print(line)

# List repetition
pattern = [1, 2, 3] * 3
print(f"\nRepeated List: {pattern}")
```

## 4. Division (/)

```python
# DIVISION (/) - Returns float result
a = 20
b = 3

result = a / b
print(f"{a} / {b} = {result}")
print(f"Result type: {type(result)}")

# Division always returns float
print(f"\n10 / 2 = {10 / 2}")  # Even exact division returns float
```

```python
# Real-world: Calculate average
scores = [85, 90, 78, 92, 88]
total = sum(scores)
count = len(scores)
average = total / count

print(f"Scores: {scores}")
print(f"Total: {total}")
print(f"Count: {count}")
print(f"Average: {average:.2f}")
```

## 5. Floor Division (//)

```python
# FLOOR DIVISION (//) - Returns integer (rounds down)
a = 20
b = 3

result = a // b
print(f"{a} // {b} = {result}")
print(f"Result type: {type(result)}")

# Comparison with regular division
print(f"\nRegular division: 17 / 5 = {17 / 5}")
print(f"Floor division: 17 // 5 = {17 // 5}")
```

```python
# Real-world: Distributing items equally
total_candies = 100
num_children = 7

candies_per_child = total_candies // num_children
leftover = total_candies % num_children

print(f"Total candies: {total_candies}")
print(f"Number of children: {num_children}")
print(f"Candies per child: {candies_per_child}")
print(f"Leftover: {leftover}")
```

## 6. Modulus (%)

```python
# MODULUS (%) - Returns remainder of division
a = 17
b = 5

result = a % b
print(f"{a} % {b} = {result} (remainder)")

# How it works: 17 = 5 * 3 + 2
#              17 // 5 = 3 (quotient)
#              17 % 5 = 2 (remainder)
```

```python
# Real-world: Check if number is even or odd
numbers = [10, 15, 22, 33, 48]

print("Checking even/odd:")
for num in numbers:
    if num % 2 == 0:
        print(f"  {num} is EVEN")
    else:
        print(f"  {num} is ODD")
```

```python
# Real-world: Check divisibility
number = 120

print(f"Divisibility of {number}:")
print(f"  Divisible by 2: {number % 2 == 0}")
print(f"  Divisible by 3: {number % 3 == 0}")
print(f"  Divisible by 5: {number % 5 == 0}")
print(f"  Divisible by 7: {number % 7 == 0}")
```

## 7. Exponentiation (**)

```python
# EXPONENTIATION (**) - Raises to power
base = 2
exponent = 10

result = base ** exponent
print(f"{base} ** {exponent} = {result}")

# Common examples
print(f"\n2 squared: 2 ** 2 = {2 ** 2}")
print(f"3 cubed: 3 ** 3 = {3 ** 3}")
print(f"Square root of 16: 16 ** 0.5 = {16 ** 0.5}")
print(f"Cube root of 27: 27 ** (1/3) = {27 ** (1/3):.2f}")
```

```python
# Real-world: Compound Interest Calculation
principal = 10000  # Initial investment
rate = 0.08        # 8% annual interest
time = 5           # 5 years
n = 12             # Compounded monthly

# Formula: A = P(1 + r/n)^(nt)
amount = principal * (1 + rate/n) ** (n * time)

print("=== Compound Interest Calculator ===")
print(f"Principal: ${principal:,}")
print(f"Annual Rate: {rate * 100}%")
print(f"Time: {time} years")
print(f"Compounded: Monthly")
print(f"Final Amount: ${amount:,.2f}")
print(f"Interest Earned: ${amount - principal:,.2f}")
```

## 8. Order of Operations (PEMDAS/BODMAS)

```python
# Order of Operations:
# 1. Parentheses ()
# 2. Exponents **
# 3. Multiplication *, Division /, Floor //, Modulus % (left to right)
# 4. Addition +, Subtraction - (left to right)

# Example 1
result1 = 2 + 3 * 4
print(f"2 + 3 * 4 = {result1}")  # 3 * 4 first, then + 2

result2 = (2 + 3) * 4
print(f"(2 + 3) * 4 = {result2}")  # Parentheses first

# Example 2
result3 = 10 + 2 ** 3 * 2
print(f"\n10 + 2 ** 3 * 2 = {result3}")  # 2**3=8, 8*2=16, 10+16=26

# Example 3
result4 = 100 / 5 / 2
print(f"100 / 5 / 2 = {result4}")  # Left to right: 100/5=20, 20/2=10
```

## 9. Complete Example: Invoice Calculator

```python
# Real-world: Invoice/Bill Calculator

# Products
laptop_price = 999.99
laptop_qty = 2

mouse_price = 29.99
mouse_qty = 5

keyboard_price = 79.99
keyboard_qty = 3

# Calculate subtotals
laptop_total = laptop_price * laptop_qty
mouse_total = mouse_price * mouse_qty
keyboard_total = keyboard_price * keyboard_qty

# Subtotal
subtotal = laptop_total + mouse_total + keyboard_total

# Discount (10%)
discount_rate = 0.10
discount_amount = subtotal * discount_rate

# After discount
after_discount = subtotal - discount_amount

# Tax (8.5%)
tax_rate = 0.085
tax_amount = after_discount * tax_rate

# Final total
total = after_discount + tax_amount

# Print invoice
print("=" * 45)
print("              TECH STORE INVOICE")
print("=" * 45)
print(f"{'Item':<20} {'Qty':>5} {'Price':>8} {'Total':>10}")
print("-" * 45)
print(f"{'Laptop':<20} {laptop_qty:>5} ${laptop_price:>7.2f} ${laptop_total:>9.2f}")
print(f"{'Mouse':<20} {mouse_qty:>5} ${mouse_price:>7.2f} ${mouse_total:>9.2f}")
print(f"{'Keyboard':<20} {keyboard_qty:>5} ${keyboard_price:>7.2f} ${keyboard_total:>9.2f}")
print("-" * 45)
print(f"{'Subtotal:':<35} ${subtotal:>8.2f}")
print(f"{'Discount (10%):':<35} -${discount_amount:>7.2f}")
print(f"{'After Discount:':<35} ${after_discount:>8.2f}")
print(f"{'Tax (8.5%):':<35} ${tax_amount:>8.2f}")
print("=" * 45)
print(f"{'TOTAL:':<35} ${total:>8.2f}")
print("=" * 45)
```

## Summary

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `10 + 3` | `13` |
| `-` | Subtraction | `10 - 3` | `7` |
| `*` | Multiplication | `10 * 3` | `30` |
| `/` | Division | `10 / 3` | `3.333...` |
| `//` | Floor Division | `10 // 3` | `3` |
| `%` | Modulus | `10 % 3` | `1` |
| `**` | Exponentiation | `10 ** 3` | `1000` |

### Key Points:
1. `/` always returns float, `//` returns integer
2. `%` is useful for checking even/odd and divisibility
3. `**` can calculate powers and roots
4. Use parentheses to control order of operations
