## 1. Basic Assignment (=)

```python
# Basic assignment
x = 10
print(f"x = {x}")

# Multiple assignment
a, b, c = 1, 2, 3
print(f"a = {a}, b = {b}, c = {c}")

# Same value to multiple variables
x = y = z = 0
print(f"x = {x}, y = {y}, z = {z}")
```

## 2. Addition Assignment (+=)

```python
# Addition Assignment (+=)
score = 100
print(f"Initial score: {score}")

score += 25  # Same as: score = score + 25
print(f"After bonus (+25): {score}")

score += 10
print(f"After another bonus (+10): {score}")
```

```python
# Real-world: Shopping Cart Total
cart_total = 0

# Adding items to cart
cart_total += 29.99  # T-shirt
print(f"Added T-shirt: ${cart_total:.2f}")

cart_total += 59.99  # Jeans
print(f"Added Jeans: ${cart_total:.2f}")

cart_total += 19.99  # Cap
print(f"Added Cap: ${cart_total:.2f}")

print(f"\nCart Total: ${cart_total:.2f}")
```

```python
# String concatenation with +=
message = "Hello"
print(f"Initial: '{message}'")

message += " "
message += "World"
message += "!"
print(f"Final: '{message}'")
```

## 3. Subtraction Assignment (-=)

```python
# Subtraction Assignment (-=)
health = 100
print(f"Initial health: {health}")

health -= 20  # Same as: health = health - 20
print(f"After damage (-20): {health}")

health -= 15
print(f"After more damage (-15): {health}")
```

```python
# Real-world: Bank Account Withdrawals
balance = 1000.00
print(f"Initial Balance: ${balance:.2f}")
print("-" * 30)

# Withdrawals
balance -= 200.00  # ATM withdrawal
print(f"ATM Withdrawal -$200.00: ${balance:.2f}")

balance -= 75.50   # Online purchase
print(f"Online Purchase -$75.50: ${balance:.2f}")

balance -= 45.00   # Subscription
print(f"Subscription -$45.00: ${balance:.2f}")

print("-" * 30)
print(f"Remaining Balance: ${balance:.2f}")
```

## 4. Multiplication Assignment (*=)

```python
# Multiplication Assignment (*=)
value = 5
print(f"Initial value: {value}")

value *= 3  # Same as: value = value * 3
print(f"After *3: {value}")

value *= 2
print(f"After *2: {value}")
```

```python
# Real-world: Investment Growth
investment = 10000
growth_rate = 1.08  # 8% annual growth

print(f"Initial Investment: ${investment:,}")
print("-" * 35)

for year in range(1, 6):
    investment *= growth_rate
    print(f"Year {year}: ${investment:,.2f}")

print("-" * 35)
print(f"After 5 years: ${investment:,.2f}")
```

## 5. Division Assignment (/=)

```python
# Division Assignment (/=)
value = 100
print(f"Initial value: {value}")

value /= 2  # Same as: value = value / 2
print(f"After /2: {value}")

value /= 5
print(f"After /5: {value}")
```

```python
# Real-world: Splitting Bill
total_bill = 150.00
num_people = 4

print(f"Total Bill: ${total_bill:.2f}")
print(f"Number of People: {num_people}")

total_bill /= num_people
print(f"Each person pays: ${total_bill:.2f}")
```

## 6. Floor Division Assignment (//=)

```python
# Floor Division Assignment (//=)
value = 100
print(f"Initial value: {value}")

value //= 3  # Same as: value = value // 3
print(f"After //3: {value}")

value //= 2
print(f"After //2: {value}")
```

```python
# Real-world: Binary Halving
number = 256
print(f"Starting number: {number}")
print("-" * 20)

steps = 0
while number > 1:
    number //= 2
    steps += 1
    print(f"Step {steps}: {number}")

print(f"\nTotal halving steps: {steps}")
```

## 7. Modulus Assignment (%=)

```python
# Modulus Assignment (%=)
value = 17
print(f"Initial value: {value}")

value %= 5  # Same as: value = value % 5
print(f"After %5: {value}")
```

```python
# Real-world: Circular Counter (like clock hours)
hour = 10

print(f"Current hour: {hour}")

# Add 5 hours
hour += 5
hour %= 12  # Keep within 1-12 range
print(f"After 5 hours: {hour if hour != 0 else 12}")

# Add 8 more hours
hour += 8
hour %= 12
print(f"After 8 more hours: {hour if hour != 0 else 12}")
```

## 8. Exponentiation Assignment (**=)

```python
# Exponentiation Assignment (**=)
value = 2
print(f"Initial value: {value}")

value **= 3  # Same as: value = value ** 3
print(f"After **3: {value}")

value **= 2
print(f"After **2: {value}")
```

```python
# Real-world: Exponential Growth (like bacteria doubling)
bacteria = 1
print(f"Initial bacteria count: {bacteria}")
print("-" * 30)

for generation in range(1, 11):
    bacteria **= 2  # Actually, this should be *= 2 for doubling
    # Let's fix this:
    pass

# Better example:
bacteria = 2
for power in range(1, 6):
    print(f"2^{power} = {bacteria ** power}")
```

## 9. Complete Example: Game Score Tracker

```python
# Real-world: Complete Game Score System

# Player stats
player_name = "Hero"
score = 0
health = 100
coins = 50
level = 1

print("=" * 40)
print("        ADVENTURE GAME")
print("=" * 40)
print(f"Player: {player_name}")
print(f"Score: {score} | Health: {health} | Coins: {coins} | Level: {level}")
print("-" * 40)

# Round 1: Defeated enemy
print("\n[Round 1] Defeated an enemy!")
score += 100
coins += 25
print(f"  Score +100, Coins +25")

# Round 2: Got hit
print("\n[Round 2] Got hit by enemy!")
health -= 30
print(f"  Health -30")

# Round 3: Found treasure
print("\n[Round 3] Found treasure chest!")
coins *= 2  # Double coins
score += 200
print(f"  Coins doubled! Score +200")

# Round 4: Bought health potion
print("\n[Round 4] Bought health potion")
coins -= 30
health += 20
print(f"  Coins -30, Health +20")

# Round 5: Level up!
print("\n[Round 5] LEVEL UP!")
level += 1
score *= 2  # Double score on level up
print(f"  Level +1, Score doubled!")

# Final stats
print("\n" + "=" * 40)
print("          FINAL STATS")
print("=" * 40)
print(f"Player: {player_name}")
print(f"Level: {level}")
print(f"Score: {score}")
print(f"Health: {health}")
print(f"Coins: {coins}")
print("=" * 40)
```

## Summary

| Operator | Example | Equivalent | Use Case |
|----------|---------|------------|----------|
| `=` | `x = 5` | - | Basic assignment |
| `+=` | `x += 3` | `x = x + 3` | Adding to total |
| `-=` | `x -= 3` | `x = x - 3` | Subtracting from balance |
| `*=` | `x *= 3` | `x = x * 3` | Scaling values |
| `/=` | `x /= 3` | `x = x / 3` | Dividing equally |
| `//=` | `x //= 3` | `x = x // 3` | Integer division |
| `%=` | `x %= 3` | `x = x % 3` | Circular counters |
| `**=` | `x **= 3` | `x = x ** 3` | Exponential growth |

### Key Points:
1. Assignment operators combine operation with assignment
2. They make code shorter and more readable
3. Commonly used in loops and accumulating values
4. Works with numbers, strings (+=), and lists (+=)
