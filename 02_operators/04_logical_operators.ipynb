## 1. AND Operator

````python
# AND - Both conditions must be True
print("=== AND TRUTH TABLE ===")
print(f"True and True   = {True and True}")
print(f"True and False  = {True and False}")
print(f"False and True  = {False and True}")
print(f"False and False = {False and False}")
````

````python
# Real-world: Login Validation
stored_username = "admin"
stored_password = "secret123"

def login(username, password):
    if username == stored_username and password == stored_password:
        print(f"‚úÖ Welcome, {username}! Login successful.")
        return True
    else:
        print("‚ùå Invalid username or password.")
        return False

# Test cases
print("Test 1: Correct credentials")
login("admin", "secret123")

print("\nTest 2: Wrong password")
login("admin", "wrongpass")

print("\nTest 3: Wrong username")
login("user", "secret123")
````

````python
# Real-world: Loan Eligibility
def check_loan_eligibility(age, credit_score, income, employment_years):
    print(f"\nChecking: Age={age}, Credit={credit_score}, Income=${income:,}, Years={employment_years}")
    
    # All conditions must be met
    if age >= 21 and credit_score >= 700 and income >= 30000 and employment_years >= 2:
        print("‚úÖ Loan APPROVED!")
        return True
    else:
        print("‚ùå Loan DENIED.")
        # Show which requirements failed
        if age < 21:
            print("   - Must be at least 21 years old")
        if credit_score < 700:
            print("   - Credit score must be 700+")
        if income < 30000:
            print("   - Income must be $30,000+")
        if employment_years < 2:
            print("   - Must have 2+ years employment")
        return False

# Test cases
check_loan_eligibility(25, 750, 50000, 3)   # Should pass
check_loan_eligibility(20, 720, 45000, 2)   # Fails age
check_loan_eligibility(30, 650, 60000, 5)   # Fails credit
````

## 2. OR Operator

````python
# OR - At least one condition must be True
print("=== OR TRUTH TABLE ===")
print(f"True or True   = {True or True}")
print(f"True or False  = {True or False}")
print(f"False or True  = {False or True}")
print(f"False or False = {False or False}")
````

````python
# Real-world: Access Control
def check_access(role):
    if role == "admin" or role == "moderator" or role == "editor":
        print(f"‚úÖ {role.upper()} has edit access!")
        return True
    else:
        print(f"‚ùå {role} does not have edit access.")
        return False

check_access("admin")
check_access("moderator")
check_access("viewer")
````

````python
# Real-world: Discount Eligibility
def check_discount(is_member, order_total, is_first_order, has_coupon):
    print(f"\nChecking: Member={is_member}, Total=${order_total}, First Order={is_first_order}, Coupon={has_coupon}")
    
    # Discount if ANY of these conditions are met
    if is_member or order_total >= 100 or is_first_order or has_coupon:
        print("üéâ You qualify for a discount!")
        
        # Show which discount applies
        if is_member:
            print("   ‚Üí Member discount: 15%")
        elif is_first_order:
            print("   ‚Üí First order discount: 10%")
        elif order_total >= 100:
            print("   ‚Üí Bulk discount: 10%")
        elif has_coupon:
            print("   ‚Üí Coupon discount: 5%")
    else:
        print("üí° No discount available. Consider joining membership!")

check_discount(True, 50, False, False)   # Member
check_discount(False, 150, False, False) # High order
check_discount(False, 30, True, False)   # First order
check_discount(False, 30, False, False)  # No discount
````

## 3. NOT Operator

````python
# NOT - Reverses the boolean value
print("=== NOT OPERATOR ===")
print(f"not True  = {not True}")
print(f"not False = {not False}")

# With variables
is_raining = True
print(f"\nis_raining = {is_raining}")
print(f"not is_raining = {not is_raining}")
````

````python
# Real-world: Feature Toggle
is_maintenance_mode = False
is_user_banned = False

def can_access_app(user):
    print(f"\nChecking access for: {user}")
    
    if not is_maintenance_mode and not is_user_banned:
        print("‚úÖ Access granted!")
        return True
    else:
        if is_maintenance_mode:
            print("üîß System under maintenance. Try again later.")
        if is_user_banned:
            print("üö´ Your account has been suspended.")
        return False

can_access_app("John")

# Simulate maintenance
is_maintenance_mode = True
can_access_app("Jane")
````

## 4. Combining Logical Operators

````python
# Combining AND, OR, NOT
print("=== COMBINED OPERATORS ===")

# Order of precedence: NOT > AND > OR
a = True
b = False
c = True

print(f"a = {a}, b = {b}, c = {c}")
print(f"\na and b or c = {a and b or c}")        # (True and False) or True = True
print(f"a or b and c = {a or b and c}")          # True or (False and True) = True
print(f"not a and b or c = {not a and b or c}")  # (not True and False) or True = True
````

````python
# Using parentheses for clarity (RECOMMENDED)
print("=== WITH PARENTHESES ===")

a = True
b = False
c = True

print(f"(a and b) or c = {(a and b) or c}")
print(f"a and (b or c) = {a and (b or c)}")
print(f"(not a) and (b or c) = {(not a) and (b or c)}")
````

## 5. Short-Circuit Evaluation

````python
# Short-circuit: Python stops evaluating as soon as result is determined

# AND: If first is False, don't check second
print("=== AND SHORT-CIRCUIT ===")

def check_a():
    print("Checking A...")
    return False

def check_b():
    print("Checking B...")
    return True

# check_b() is never called because check_a() is False
result = check_a() and check_b()
print(f"Result: {result}")
````

````python
# OR: If first is True, don't check second
print("=== OR SHORT-CIRCUIT ===")

def check_a():
    print("Checking A...")
    return True

def check_b():
    print("Checking B...")
    return False

# check_b() is never called because check_a() is True
result = check_a() or check_b()
print(f"Result: {result}")
````

````python
# Practical use: Safe dictionary access
user = {"name": "John", "age": 25}

# Check if key exists AND has valid value
if "email" in user and len(user["email"]) > 0:
    print(f"Email: {user['email']}")
else:
    print("Email not provided")

# Safe division
x = 10
y = 0

if y != 0 and x / y > 1:
    print("Division result > 1")
else:
    print("Cannot divide or result <= 1")
````

## 6. Complete Example: E-commerce Checkout

````python
# Real-world: Complete E-commerce Checkout System

def process_checkout(user_logged_in, cart_items, payment_valid, address_valid, 
                     is_member, order_total, has_coupon):
    print("="*50)
    print("        CHECKOUT PROCESSING")
    print("="*50)
    
    # Step 1: Check if user can checkout
    print("\nüìã Step 1: Validating checkout requirements...")
    
    if not user_logged_in:
        print("‚ùå Please log in to continue.")
        return False
    
    if not cart_items:  # Empty cart
        print("‚ùå Your cart is empty.")
        return False
    
    if not payment_valid or not address_valid:
        if not payment_valid:
            print("‚ùå Invalid payment method.")
        if not address_valid:
            print("‚ùå Invalid shipping address.")
        return False
    
    print("‚úÖ All requirements met!")
    
    # Step 2: Calculate discounts
    print("\nüí∞ Step 2: Calculating discounts...")
    discount = 0
    
    if is_member and order_total >= 100:
        discount = 20
        print(f"   ‚Üí VIP Member with $100+ order: {discount}% off!")
    elif is_member or has_coupon:
        discount = 10
        reason = "Member discount" if is_member else "Coupon applied"
        print(f"   ‚Üí {reason}: {discount}% off!")
    elif order_total >= 50:
        discount = 5
        print(f"   ‚Üí Order over $50: {discount}% off!")
    else:
        print("   ‚Üí No discounts applied.")
    
    # Step 3: Calculate final price
    print("\nüßæ Step 3: Order Summary")
    print(f"   Subtotal: ${order_total:.2f}")
    discount_amount = order_total * (discount / 100)
    final_total = order_total - discount_amount
    print(f"   Discount ({discount}%): -${discount_amount:.2f}")
    print(f"   Final Total: ${final_total:.2f}")
    
    # Step 4: Shipping
    print("\nüì¶ Step 4: Shipping...")
    if is_member or final_total >= 75:
        print("   ‚úÖ FREE shipping!")
    else:
        print("   üì¶ Standard shipping: $5.99")
        final_total += 5.99
    
    print("\n" + "="*50)
    print(f"    ORDER CONFIRMED! Total: ${final_total:.2f}")
    print("="*50)
    return True

# Test successful checkout
process_checkout(
    user_logged_in=True,
    cart_items=["Laptop", "Mouse"],
    payment_valid=True,
    address_valid=True,
    is_member=True,
    order_total=150.00,
    has_coupon=False
)
````

````python
# Test failed checkout
process_checkout(
    user_logged_in=True,
    cart_items=["Book"],
    payment_valid=False,  # Invalid payment
    address_valid=True,
    is_member=False,
    order_total=25.00,
    has_coupon=True
)
````

## Summary

| Operator | Description | Example | Result |
|----------|-------------|---------|--------|
| `and` | Both True | `True and True` | `True` |
| `or` | At least one True | `True or False` | `True` |
| `not` | Reverse boolean | `not True` | `False` |

### Truth Tables:
```
AND:              OR:               NOT:
T and T = T       T or T = T        not T = F
T and F = F       T or F = T        not F = T
F and T = F       F or T = T
F and F = F       F or F = F
```

### Key Points:
1. Precedence: `not` > `and` > `or`
2. Use parentheses for clarity
3. Short-circuit evaluation can improve efficiency
4. Empty collections are `False` in boolean context
