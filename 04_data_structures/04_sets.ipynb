## 1. Creating Sets

```python
# Different ways to create sets

# Using curly braces (NOT empty - that creates a dict!)
fruits = {"apple", "banana", "cherry"}

# Empty set (must use set())
empty = set()  # NOT {} - that's an empty dict!

# From a list (removes duplicates)
numbers = set([1, 2, 2, 3, 3, 3, 4])

# From a string (unique characters)
letters = set("hello")

# Mixed types (must be immutable)
mixed = {1, "hello", 3.14, (1, 2)}

print(f"Fruits: {fruits}")
print(f"Empty: {empty}, type: {type(empty)}")
print(f"Numbers (duplicates removed): {numbers}")
print(f"Letters: {letters}")
print(f"Mixed: {mixed}")
```

```python
# Sets are unordered - no indexing!
my_set = {1, 2, 3, 4, 5}

try:
    print(my_set[0])  # This fails!
except TypeError as e:
    print(f"Error: {e}")
    print("Sets don't support indexing - they're unordered!")
```

## 2. Adding and Removing Elements

```python
# add() - Add single element
fruits = {"apple", "banana"}
print(f"Original: {fruits}")

fruits.add("cherry")
print(f"After add('cherry'): {fruits}")

# Adding duplicate has no effect
fruits.add("apple")
print(f"After add('apple'): {fruits}")
```

```python
# update() - Add multiple elements
fruits = {"apple", "banana"}

fruits.update(["cherry", "date"])  # From list
print(f"After update(list): {fruits}")

fruits.update({"elderberry"}, ("fig",))  # Multiple iterables
print(f"After update(set, tuple): {fruits}")
```

```python
# remove() - Remove element (raises error if not found)
fruits = {"apple", "banana", "cherry"}

fruits.remove("banana")
print(f"After remove('banana'): {fruits}")

try:
    fruits.remove("grape")
except KeyError as e:
    print(f"KeyError: {e}")
```

```python
# discard() - Remove element (no error if not found)
fruits = {"apple", "banana", "cherry"}

fruits.discard("banana")
print(f"After discard('banana'): {fruits}")

fruits.discard("grape")  # No error!
print(f"After discard('grape'): {fruits}")
```

```python
# pop() - Remove and return arbitrary element
fruits = {"apple", "banana", "cherry"}

popped = fruits.pop()
print(f"Popped: {popped}")
print(f"Remaining: {fruits}")

# clear() - Remove all elements
fruits.clear()
print(f"After clear: {fruits}")
```

## 3. Set Operations

```python
# Union - All elements from both sets
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Using | operator
union1 = set1 | set2

# Using union() method
union2 = set1.union(set2)

print(f"Set1: {set1}")
print(f"Set2: {set2}")
print(f"Union: {union1}")
```

```python
# Intersection - Elements in both sets
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Using & operator
intersection1 = set1 & set2

# Using intersection() method
intersection2 = set1.intersection(set2)

print(f"Set1: {set1}")
print(f"Set2: {set2}")
print(f"Intersection: {intersection1}")
```

```python
# Difference - Elements in first but not in second
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Using - operator
diff1 = set1 - set2  # In set1 but not in set2
diff2 = set2 - set1  # In set2 but not in set1

print(f"Set1: {set1}")
print(f"Set2: {set2}")
print(f"Set1 - Set2: {diff1}")
print(f"Set2 - Set1: {diff2}")
```

```python
# Symmetric Difference - Elements in either set but not both
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Using ^ operator
sym_diff = set1 ^ set2

print(f"Set1: {set1}")
print(f"Set2: {set2}")
print(f"Symmetric Difference: {sym_diff}")
```

## 4. Set Comparison Methods

```python
# Subset and Superset
set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}
set3 = {1, 2, 3}

print(f"Set1: {set1}")
print(f"Set2: {set2}")
print(f"Set3: {set3}")

print(f"\nset1 is subset of set2: {set1.issubset(set2)}")
print(f"set2 is superset of set1: {set2.issuperset(set1)}")
print(f"set1 == set3: {set1 == set3}")

# Using operators
print(f"\nset1 <= set2 (subset): {set1 <= set2}")
print(f"set1 < set2 (proper subset): {set1 < set2}")
print(f"set1 < set3 (proper subset): {set1 < set3}")  # False, they're equal
```

```python
# Disjoint - No common elements
set1 = {1, 2, 3}
set2 = {4, 5, 6}
set3 = {3, 4, 5}

print(f"Set1: {set1}")
print(f"Set2: {set2}")
print(f"Set3: {set3}")

print(f"\nset1 and set2 are disjoint: {set1.isdisjoint(set2)}")
print(f"set1 and set3 are disjoint: {set1.isdisjoint(set3)}")
```

## 5. Practical Examples

```python
# Example 1: Remove duplicates from a list
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
unique = list(set(numbers))

print(f"Original: {numbers}")
print(f"Unique: {unique}")
```

```python
# Example 2: Find common friends
alice_friends = {"Bob", "Charlie", "Diana", "Eve"}
bob_friends = {"Alice", "Charlie", "Frank", "Eve"}

# Common friends
common = alice_friends & bob_friends
print(f"Common friends: {common}")

# Friends of Alice but not Bob
alice_only = alice_friends - bob_friends
print(f"Alice's friends (not Bob's): {alice_only}")

# All friends
all_friends = alice_friends | bob_friends
print(f"All friends: {all_friends}")
```

```python
# Example 3: Unique words in text
text = "the quick brown fox jumps over the lazy dog"
words = text.split()
unique_words = set(words)

print(f"Total words: {len(words)}")
print(f"Unique words: {len(unique_words)}")
print(f"Words: {unique_words}")
```

```python
# Example 4: Membership testing (fast!)
valid_users = {"alice", "bob", "charlie", "diana"}

def check_access(username):
    if username.lower() in valid_users:
        return f"âœ… {username} has access"
    return f"âŒ {username} denied"

print(check_access("Alice"))
print(check_access("Eve"))
```

## 6. Frozen Sets

```python
# Frozen sets are immutable sets
frozen = frozenset([1, 2, 3, 4, 5])
print(f"Frozen set: {frozen}")

# Cannot modify
try:
    frozen.add(6)
except AttributeError as e:
    print(f"Error: {e}")
```

```python
# Frozen sets can be used as dictionary keys or set elements
fs1 = frozenset([1, 2])
fs2 = frozenset([3, 4])

# Set of frozen sets
set_of_sets = {fs1, fs2}
print(f"Set of frozen sets: {set_of_sets}")

# As dictionary key
permissions = {
    frozenset(["read"]): "viewer",
    frozenset(["read", "write"]): "editor",
    frozenset(["read", "write", "admin"]): "admin"
}

print(f"\nPermissions dict: {permissions}")
```

## 7. Complete Example: Course Enrollment System

```python
class CourseEnrollment:
    def __init__(self):
        self.courses = {}  # course_name: set of students
    
    def add_course(self, course_name):
        if course_name not in self.courses:
            self.courses[course_name] = set()
            print(f"âœ… Added course: {course_name}")
        else:
            print(f"Course {course_name} already exists")
    
    def enroll(self, student, course_name):
        if course_name in self.courses:
            self.courses[course_name].add(student)
            print(f"âœ… {student} enrolled in {course_name}")
        else:
            print(f"âŒ Course {course_name} not found")
    
    def drop(self, student, course_name):
        if course_name in self.courses:
            self.courses[course_name].discard(student)
            print(f"âœ… {student} dropped from {course_name}")
    
    def get_common_students(self, course1, course2):
        """Students enrolled in both courses."""
        return self.courses.get(course1, set()) & self.courses.get(course2, set())
    
    def get_all_students(self, *course_names):
        """Students enrolled in any of the courses."""
        result = set()
        for course in course_names:
            result |= self.courses.get(course, set())
        return result
    
    def get_exclusive_students(self, course_name):
        """Students enrolled ONLY in this course."""
        all_other = set()
        for name, students in self.courses.items():
            if name != course_name:
                all_other |= students
        return self.courses.get(course_name, set()) - all_other
    
    def display(self):
        print("\n" + "="*50)
        print("           COURSE ENROLLMENT REPORT")
        print("="*50)
        for course, students in self.courses.items():
            print(f"\nğŸ“š {course} ({len(students)} students):")
            for student in sorted(students):
                print(f"   â€¢ {student}")
        print("="*50)


# Demo
system = CourseEnrollment()

# Add courses
for course in ["Python", "JavaScript", "Data Science"]:
    system.add_course(course)

# Enroll students
enrollments = [
    ("Alice", "Python"), ("Alice", "JavaScript"), ("Alice", "Data Science"),
    ("Bob", "Python"), ("Bob", "JavaScript"),
    ("Charlie", "Python"),
    ("Diana", "JavaScript"), ("Diana", "Data Science"),
    ("Eve", "Data Science")
]

print("\n--- Enrolling Students ---")
for student, course in enrollments:
    system.enroll(student, course)

# Display all
system.display()

# Analytics
print("\nğŸ“Š ANALYTICS:")
print(f"Students in both Python & JavaScript: {system.get_common_students('Python', 'JavaScript')}")
print(f"All CS students: {system.get_all_students('Python', 'JavaScript')}")
print(f"Students ONLY in Data Science: {system.get_exclusive_students('Data Science')}")
```

## Summary

### Set Operations Quick Reference:

| Operation | Operator | Method | Description |
|-----------|----------|--------|-------------|
| Union | `\|` | `union()` | All elements |
| Intersection | `&` | `intersection()` | Common elements |
| Difference | `-` | `difference()` | In first, not second |
| Symmetric Diff | `^` | `symmetric_difference()` | Not in both |

### Set Methods:

| Method | Description |
|--------|-------------|
| `add(x)` | Add element |
| `remove(x)` | Remove (raises error if missing) |
| `discard(x)` | Remove (no error if missing) |
| `pop()` | Remove arbitrary element |
| `issubset()` | Check if subset |
| `issuperset()` | Check if superset |
| `isdisjoint()` | Check if no common elements |

### Key Points:
1. Sets contain **unique** elements only
2. Sets are **unordered** - no indexing
3. Elements must be **immutable**
4. Membership testing is **O(1)** - very fast!
5. Use **frozenset** for immutable sets
