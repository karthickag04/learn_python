```python
# make examples for all the comparison operators
# ==, !=, >, <, >=, <=

x = 10
y = 20

if x == y:
    print("x is equal to y")
else:
    print("x is not equal to y")

if x != y:
    print("x is not equal to y")
else:
    print("x is equal to y")

if x > y:
    print("x is greater than y")
else:
    print("x is not greater than y")

if x < y:
    print("x is less than y")
else:
    print("x is not less than y")

if x >= y:
    print("x is greater than or equal to y")
else:
    print("x is not greater than or equal to y")

if x <= y:
    print("x is less than or equal to y")
else:
    print("x is not less than or equal to y")
```

```python
# make examples for all the comparison operators with logical operators
# ==, !=, >, <, >=, <=

x = 10
y = 20
z = 15

if x == 10 and y == 20 and z == 15:
    print("x is equal to 10 and y is equal to 20")

if x != y and x < y:
    print("x is not equal to y and x is less than y")

if x > z or y > z:
    print("either x is greater than z or y is greater than z")

if x < z or y < z:
    print("either x is less than z or y is less than z")

if x >= 10 and y >= 10:
    print("x and y are both greater than or equal to 10")

if not x == y and x <= y:
    print("x is not equal to y and x is less than or equal to y")
```

```python
# match usecase
day = "friday"

match day:
    case 9:
        print("Start of week 😅")
    case "tuesday":
        print("Getting going 📈")
    case "wednesday":
        print("Midweek hump 🐪")
    case "thursday":
        print("Almost there! 🏃")
    case "friday":
        print("TGIF! 🎉")
    case "saturday":
        print("Weekend party! 🕺")
    case "sunday":
        print("Rest day 😴")
    case _:
        print("Invalid day")
```

```python
def operation(x, y, operator):
    if operator == "+":
        return x + y
    elif operator == "-":
        return x - y
    elif operator == "*":
        return x * y
    elif operator == "/":
        return x / y
    else:
        return "Invalid operator"
```

```python
operation(10, 20, "+")
operation(10, 20, "-")
```
