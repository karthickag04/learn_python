## 1. Basic Match-Case

```python
# Basic syntax
def get_day_type(day):
    match day.lower():
        case "monday":
            return "Start of week ğŸ˜…"
        case "tuesday":
            return "Getting going ğŸ“ˆ"
        case "wednesday":
            return "Midweek hump ğŸª"
        case "thursday":
            return "Almost there! ğŸƒ"
        case "friday":
            return "TGIF! ğŸ‰"
        case "saturday":
            return "Weekend party! ğŸ•º"
        case "sunday":
            return "Rest day ğŸ˜´"
        case _:  # Default case (wildcard)
            return "Invalid day"

print(get_day_type("Monday"))
print(get_day_type("Friday"))
print(get_day_type("Funday"))
```

```python
# Real-world: HTTP Status Codes
def get_status_message(status_code):
    match status_code:
        case 200:
            return "âœ… OK - Request successful"
        case 201:
            return "âœ… Created - Resource created"
        case 301:
            return "â†—ï¸ Moved Permanently"
        case 400:
            return "âŒ Bad Request - Check your input"
        case 401:
            return "ğŸ”’ Unauthorized - Login required"
        case 403:
            return "ğŸš« Forbidden - Access denied"
        case 404:
            return "ğŸ” Not Found - Resource doesn't exist"
        case 500:
            return "ğŸ’¥ Internal Server Error"
        case _:
            return f"Unknown status code: {status_code}"

codes = [200, 404, 500, 418]
for code in codes:
    print(f"{code}: {get_status_message(code)}")
```

## 2. Multiple Patterns (OR patterns)

```python
# Use | to match multiple patterns
def get_day_category(day):
    match day.lower():
        case "saturday" | "sunday":
            return "ğŸ–ï¸ Weekend"
        case "monday" | "tuesday" | "wednesday" | "thursday" | "friday":
            return "ğŸ’¼ Weekday"
        case _:
            return "âŒ Invalid day"

print(get_day_category("Monday"))
print(get_day_category("Saturday"))
```

```python
# Real-world: File type classifier
def classify_file(extension):
    match extension.lower():
        case ".jpg" | ".jpeg" | ".png" | ".gif" | ".bmp" | ".webp":
            return "ğŸ–¼ï¸ Image file"
        case ".mp4" | ".avi" | ".mkv" | ".mov" | ".wmv":
            return "ğŸ¬ Video file"
        case ".mp3" | ".wav" | ".flac" | ".aac" | ".ogg":
            return "ğŸµ Audio file"
        case ".pdf" | ".doc" | ".docx" | ".txt" | ".odt":
            return "ğŸ“„ Document file"
        case ".py" | ".js" | ".java" | ".cpp" | ".html" | ".css":
            return "ğŸ’» Code file"
        case ".zip" | ".rar" | ".7z" | ".tar" | ".gz":
            return "ğŸ“¦ Archive file"
        case _:
            return "â“ Unknown file type"

files = [".py", ".jpg", ".mp3", ".pdf", ".xyz"]
for ext in files:
    print(f"{ext}: {classify_file(ext)}")
```

## 3. Guard Clauses (if conditions)

```python
# Add conditions with 'if' after pattern
def categorize_number(num):
    match num:
        case n if n < 0:
            return "Negative"
        case 0:
            return "Zero"
        case n if n < 10:
            return "Single digit"
        case n if n < 100:
            return "Double digit"
        case n if n < 1000:
            return "Triple digit"
        case _:
            return "Very large"

numbers = [-5, 0, 7, 42, 256, 10000]
for num in numbers:
    print(f"{num:6d}: {categorize_number(num)}")
```

```python
# Real-world: Age-based pricing
def get_ticket_price(age):
    match age:
        case a if a < 0:
            return "Invalid age"
        case a if a < 3:
            return "Free (infant)"
        case a if a < 12:
            return "$5.00 (child)"
        case a if a < 18:
            return "$8.00 (teen)"
        case a if a < 65:
            return "$12.00 (adult)"
        case _:
            return "$6.00 (senior)"

ages = [2, 8, 15, 30, 70, -1]
for age in ages:
    print(f"Age {age:2d}: {get_ticket_price(age)}")
```

## 4. Pattern Matching with Sequences

```python
# Match lists and tuples
def describe_point(point):
    match point:
        case (0, 0):
            return "Origin"
        case (0, y):
            return f"On Y-axis at y={y}"
        case (x, 0):
            return f"On X-axis at x={x}"
        case (x, y) if x == y:
            return f"On diagonal at ({x}, {y})"
        case (x, y):
            return f"Point at ({x}, {y})"
        case _:
            return "Not a valid point"

points = [(0, 0), (5, 0), (0, 3), (4, 4), (2, 5), (1, 2, 3)]
for p in points:
    print(f"{str(p):12s} â†’ {describe_point(p)}")
```

```python
# Match variable-length sequences
def process_command(command):
    match command.split():
        case ["quit"] | ["exit"]:
            return "ğŸ‘‹ Goodbye!"
        case ["hello"]:
            return "ğŸ‘‹ Hello! How can I help?"
        case ["hello", name]:
            return f"ğŸ‘‹ Hello, {name}!"
        case ["add", x, y]:
            return f"ğŸ“Š {x} + {y} = {int(x) + int(y)}"
        case ["search", *keywords]:
            return f"ğŸ” Searching for: {' '.join(keywords)}"
        case []:
            return "â“ Please enter a command"
        case _:
            return "âŒ Unknown command"

commands = [
    "hello",
    "hello Alice",
    "add 5 3",
    "search python programming tutorial",
    "quit",
    "",
    "unknown command here"
]

for cmd in commands:
    print(f"'{cmd}' â†’ {process_command(cmd)}")
```

## 5. Pattern Matching with Dictionaries

```python
# Match dictionary patterns
def process_user_action(action):
    match action:
        case {"type": "login", "username": user}:
            return f"ğŸ”‘ User {user} logged in"
        case {"type": "logout", "username": user}:
            return f"ğŸ‘‹ User {user} logged out"
        case {"type": "purchase", "item": item, "price": price}:
            return f"ğŸ›’ Purchased {item} for ${price}"
        case {"type": "message", "to": recipient, "content": msg}:
            return f"âœ‰ï¸ Message to {recipient}: {msg}"
        case _:
            return "â“ Unknown action"

actions = [
    {"type": "login", "username": "alice"},
    {"type": "purchase", "item": "laptop", "price": 999},
    {"type": "message", "to": "bob", "content": "Hello!"},
    {"type": "logout", "username": "alice"},
    {"type": "unknown"}
]

for action in actions:
    print(process_user_action(action))
```

## 6. Complete Example: Command-Line Interface

```python
# Complete CLI Application
class SimpleCLI:
    def __init__(self):
        self.running = True
        self.tasks = []
        self.username = "Guest"
    
    def process(self, command):
        parts = command.strip().split()
        
        match parts:
            # Exit commands
            case ["exit"] | ["quit"] | ["q"]:
                self.running = False
                return "ğŸ‘‹ Goodbye!"
            
            # Help command
            case ["help"]:
                return """ğŸ“š Available Commands:
  help              - Show this help
  user <name>       - Set username
  add <task>        - Add a task
  list              - List all tasks
  done <number>     - Mark task as done
  clear             - Clear all tasks
  exit              - Exit program"""
            
            # Set username
            case ["user", name]:
                self.username = name
                return f"ğŸ‘¤ Username set to: {name}"
            
            # Add task
            case ["add", *task_words] if task_words:
                task = " ".join(task_words)
                self.tasks.append({"text": task, "done": False})
                return f"âœ… Added task: {task}"
            
            # List tasks
            case ["list"]:
                if not self.tasks:
                    return "ğŸ“‹ No tasks yet. Use 'add <task>' to add one."
                result = f"ğŸ“‹ Tasks for {self.username}:\n"
                for i, task in enumerate(self.tasks, 1):
                    status = "âœ…" if task["done"] else "â¬œ"
                    result += f"  {i}. {status} {task['text']}\n"
                return result.strip()
            
            # Mark task done
            case ["done", num] if num.isdigit():
                idx = int(num) - 1
                if 0 <= idx < len(self.tasks):
                    self.tasks[idx]["done"] = True
                    return f"âœ… Marked task {num} as done!"
                return f"âŒ Task {num} not found."
            
            # Clear tasks
            case ["clear"]:
                count = len(self.tasks)
                self.tasks = []
                return f"ğŸ—‘ï¸ Cleared {count} tasks."
            
            # Empty input
            case []:
                return ""
            
            # Unknown command
            case _:
                return f"â“ Unknown command: {' '.join(parts)}\n   Type 'help' for available commands."

# Demo the CLI
cli = SimpleCLI()

demo_commands = [
    "help",
    "user Alice",
    "add Buy groceries",
    "add Complete Python lesson",
    "add Call mom",
    "list",
    "done 2",
    "list",
    "unknown command",
]

print("="*50)
print("     SIMPLE CLI DEMO")
print("="*50)

for cmd in demo_commands:
    print(f"\n> {cmd}")
    result = cli.process(cmd)
    if result:
        print(result)
```

## Comparison: if-elif-else vs match-case

```python
# Same logic, two ways

# Using if-elif-else (works in all Python versions)
def grade_if(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# Using match-case (Python 3.10+)
def grade_match(score):
    match score:
        case s if s >= 90:
            return "A"
        case s if s >= 80:
            return "B"
        case s if s >= 70:
            return "C"
        case s if s >= 60:
            return "D"
        case _:
            return "F"

# Both produce same results
for score in [95, 82, 73, 65, 45]:
    print(f"Score {score}: if-elif={grade_if(score)}, match={grade_match(score)}")
```

## Summary

| Feature | Syntax | Example |
|---------|--------|--------|
| Basic match | `case value:` | `case "hello":` |
| OR patterns | `case val1 \| val2:` | `case "y" \| "yes":` |
| Wildcard | `case _:` | Default case |
| Guard clause | `case x if cond:` | `case n if n > 0:` |
| Capture value | `case x:` | Variable binding |
| Sequence match | `case [a, b]:` | Match list/tuple |
| Variable length | `case [first, *rest]:` | Match with rest |
| Dict match | `case {"key": val}:` | Match dictionary |

### When to Use match-case:
1. âœ… Multiple discrete values to check
2. âœ… Pattern matching with data structures
3. âœ… Command/action processing
4. âŒ Simple boolean conditions â†’ use if-else
5. âŒ Range checks â†’ if-elif often clearer
