## 1. Basic `if` Statement

```python
# Simple if statement
# Syntax: if condition:
#             code to execute

age = 18

if age >= 18:
    print("You are eligible to vote!")

print("Program continues...")
```

```python
# Multiple conditions in same if block
temperature = 35

if temperature > 30:
    print("üî• It's hot today!")
    print("üíß Stay hydrated!")
    print("üè† Prefer staying indoors.")
```

```python
# Real-world: Simple validation
username = "admin"

if username:
    print(f"Welcome, {username}!")

# Check for empty username
empty_username = ""

if not empty_username:
    print("Username cannot be empty!")
```

## 2. `if-else` Statement

```python
# if-else: Execute one block or the other

age = 16

if age >= 18:
    print("‚úÖ You are an adult.")
else:
    print("‚ùå You are a minor.")
```

```python
# Real-world: Login system
def login(username, password):
    stored_username = "admin"
    stored_password = "secret123"
    
    if username == stored_username and password == stored_password:
        print(f"‚úÖ Welcome, {username}! Login successful.")
        return True
    else:
        print("‚ùå Invalid credentials. Please try again.")
        return False

# Test
login("admin", "secret123")
login("admin", "wrong")
```

```python
# Real-world: Even or Odd checker
def check_even_odd(number):
    if number % 2 == 0:
        print(f"{number} is EVEN")
    else:
        print(f"{number} is ODD")

check_even_odd(10)
check_even_odd(7)
check_even_odd(0)
```

## 3. `if-elif-else` Chain

```python
# elif: Check multiple conditions

score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} ‚Üí Grade: {grade}")
```

```python
# Real-world: Complete grading system
def calculate_grade(score):
    print(f"\nScore: {score}/100")
    
    if score < 0 or score > 100:
        print("‚ùå Invalid score! Must be 0-100.")
        return None
    elif score >= 90:
        grade = "A+"
        message = "üåü Outstanding!"
    elif score >= 80:
        grade = "A"
        message = "‚≠ê Excellent!"
    elif score >= 70:
        grade = "B"
        message = "üëç Good job!"
    elif score >= 60:
        grade = "C"
        message = "üòä Satisfactory"
    elif score >= 50:
        grade = "D"
        message = "üí™ Needs improvement"
    else:
        grade = "F"
        message = "üìö Please retake the exam"
    
    print(f"Grade: {grade}")
    print(f"Feedback: {message}")
    return grade

# Test different scores
calculate_grade(95)
calculate_grade(75)
calculate_grade(45)
calculate_grade(105)
```

```python
# Real-world: Day of week
def get_day_name(day_number):
    if day_number == 1:
        return "Monday"
    elif day_number == 2:
        return "Tuesday"
    elif day_number == 3:
        return "Wednesday"
    elif day_number == 4:
        return "Thursday"
    elif day_number == 5:
        return "Friday"
    elif day_number == 6:
        return "Saturday"
    elif day_number == 7:
        return "Sunday"
    else:
        return "Invalid day number"

print(get_day_name(1))
print(get_day_name(5))
print(get_day_name(8))
```

## 4. Nested Conditions

```python
# Nested if: Conditions inside conditions

age = 25
has_license = True

if age >= 18:
    print("Age verified: 18+")
    if has_license:
        print("‚úÖ You can drive!")
    else:
        print("‚ö†Ô∏è You need a license to drive.")
else:
    print("‚ùå You must be 18+ to drive.")
```

```python
# Real-world: ATM Withdrawal System
def atm_withdrawal(balance, amount, pin_entered, correct_pin):
    print(f"\n{'='*40}")
    print("        ATM WITHDRAWAL")
    print(f"{'='*40}")
    print(f"Current Balance: ${balance:,.2f}")
    print(f"Withdrawal Amount: ${amount:,.2f}")
    
    if pin_entered == correct_pin:
        print("‚úÖ PIN verified")
        
        if amount <= 0:
            print("‚ùå Invalid amount. Must be positive.")
        elif amount > balance:
            print("‚ùå Insufficient balance!")
            print(f"   Available: ${balance:,.2f}")
        elif amount > 500:
            print("‚ùå Daily limit exceeded!")
            print("   Maximum: $500 per transaction")
        else:
            new_balance = balance - amount
            print(f"‚úÖ Withdrawal successful!")
            print(f"   Dispensed: ${amount:,.2f}")
            print(f"   New Balance: ${new_balance:,.2f}")
            return new_balance
    else:
        print("‚ùå Invalid PIN! Transaction cancelled.")
    
    return balance

# Test scenarios
balance = 1000
balance = atm_withdrawal(balance, 200, "1234", "1234")  # Success
balance = atm_withdrawal(balance, 600, "1234", "1234")  # Over limit
balance = atm_withdrawal(balance, 100, "0000", "1234")  # Wrong PIN
```

```python
# Real-world: Movie ticket pricing
def calculate_ticket_price(age, is_student, is_weekend, is_3d):
    print(f"\nüé¨ Movie Ticket Price Calculator")
    print(f"   Age: {age}, Student: {is_student}, Weekend: {is_weekend}, 3D: {is_3d}")
    
    base_price = 12.00
    
    # Age-based discount
    if age < 12:
        price = base_price * 0.50  # 50% off for kids
        category = "Child"
    elif age >= 65:
        price = base_price * 0.70  # 30% off for seniors
        category = "Senior"
    else:
        price = base_price
        category = "Adult"
        
        # Student discount only for adults
        if is_student:
            price *= 0.85  # 15% student discount
            category = "Student"
    
    # Weekend surcharge
    if is_weekend:
        price += 3.00
    
    # 3D surcharge
    if is_3d:
        price += 4.00
    
    print(f"   Category: {category}")
    print(f"   Final Price: ${price:.2f}")
    return price

# Test different scenarios
calculate_ticket_price(8, False, False, False)   # Child
calculate_ticket_price(70, False, True, True)    # Senior + Weekend + 3D
calculate_ticket_price(22, True, False, False)   # Student
calculate_ticket_price(35, False, True, True)    # Adult + Weekend + 3D
```

## 5. Ternary Operator (Conditional Expression)

```python
# Ternary: value_if_true if condition else value_if_false

age = 20

# Traditional way
if age >= 18:
    status = "Adult"
else:
    status = "Minor"

print(f"Traditional: {status}")

# Ternary way (one line)
status = "Adult" if age >= 18 else "Minor"
print(f"Ternary: {status}")
```

```python
# More ternary examples

# Check even/odd
number = 7
result = "Even" if number % 2 == 0 else "Odd"
print(f"{number} is {result}")

# Get max value
a = 10
b = 25
max_val = a if a > b else b
print(f"Max of {a} and {b} is {max_val}")

# Sign check
num = -5
sign = "Positive" if num > 0 else "Negative" if num < 0 else "Zero"
print(f"{num} is {sign}")
```

```python
# Real-world: Greeting based on time
def get_greeting(hour):
    return (
        "Good Morning! üåÖ" if 5 <= hour < 12 else
        "Good Afternoon! ‚òÄÔ∏è" if 12 <= hour < 17 else
        "Good Evening! üåÜ" if 17 <= hour < 21 else
        "Good Night! üåô"
    )

for hour in [7, 13, 18, 23]:
    print(f"{hour}:00 ‚Üí {get_greeting(hour)}")
```

## 6. Complete Example: Student Management System

```python
# Complete Student Admission System

def check_admission(name, age, marks, has_sports_quota, has_reservation):
    print("\n" + "="*50)
    print("     STUDENT ADMISSION SYSTEM")
    print("="*50)
    print(f"Applicant: {name}")
    print(f"Age: {age} | Marks: {marks}%")
    print(f"Sports Quota: {has_sports_quota} | Reservation: {has_reservation}")
    print("-"*50)
    
    # Step 1: Age eligibility
    if age < 16:
        print("‚ùå REJECTED: Minimum age is 16 years.")
        return False
    elif age > 25:
        print("‚ö†Ô∏è Age Relaxation Required: Applicant is over 25.")
    else:
        print("‚úÖ Age Criteria: Met")
    
    # Step 2: Calculate adjusted marks
    adjusted_marks = marks
    
    if has_sports_quota:
        adjusted_marks += 5
        print(f"‚úÖ Sports Quota: +5% bonus (Adjusted: {adjusted_marks}%)")
    
    if has_reservation:
        adjusted_marks += 3
        print(f"‚úÖ Reservation: +3% bonus (Adjusted: {adjusted_marks}%)")
    
    # Step 3: Determine admission status
    print("-"*50)
    
    if adjusted_marks >= 90:
        print("üåü ADMITTED with SCHOLARSHIP!")
        print("   Category: Merit List")
        fee_waiver = 100
    elif adjusted_marks >= 80:
        print("‚≠ê ADMITTED to First Round!")
        print("   Category: First Merit")
        fee_waiver = 50
    elif adjusted_marks >= 70:
        print("‚úÖ ADMITTED to Second Round!")
        print("   Category: Second Merit")
        fee_waiver = 25
    elif adjusted_marks >= 60:
        print("üìã WAITLISTED")
        print("   Will be contacted if seats available.")
        fee_waiver = 0
    else:
        print("‚ùå NOT ADMITTED")
        print("   Minimum 60% required for admission.")
        return False
    
    if fee_waiver > 0:
        print(f"   Fee Waiver: {fee_waiver}%")
    
    print("="*50)
    return True

# Test different scenarios
check_admission("Alice", 18, 92, False, False)  # Merit with scholarship
check_admission("Bob", 17, 75, True, True)      # With quotas
check_admission("Charlie", 15, 85, False, False) # Too young
check_admission("Diana", 20, 55, False, False)  # Low marks
```

## Summary

| Statement | Use Case |
|-----------|----------|
| `if` | Single condition check |
| `if-else` | Either/or decision |
| `if-elif-else` | Multiple conditions |
| Nested `if` | Conditions within conditions |
| Ternary | One-line conditional assignment |

### Key Points:
1. Indentation is crucial in Python (4 spaces or 1 tab)
2. Use `elif` instead of multiple `if` for mutually exclusive conditions
3. Ternary is great for simple conditions, not complex logic
4. Order matters - Python checks conditions from top to bottom
