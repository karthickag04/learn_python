```python
def operations_with_input(a, b, op):
    if op == "+":
        result = a + b
        print(f"The result of {a} + {b} is: {result}")
    elif op == "-":
        result = a - b
        print(f"The result of {a} - {b} is: {result}")
    elif op == "*":
        result = a * b
        print(f"The result of {a} * {b} is: {result}")
    elif op == "/":
        if b != 0:
            result = a / b
            print(f"The result of {a} / {b} is: {result}")
        else:
            print("Error: Division by zero is not allowed.")
    else:
        print("Error: Invalid operation. Please choose +, -, *, or /.")
```

```python
def operations_with_input_return(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        if b != 0:
            return a / b
        else:
            return "Error: Division by zero is not allowed."
    else:
        return "Error: Invalid operation. Please choose +, -, *, or /."
```

```python
operations_with_input(10, 5, "+")
operations_with_input(10, 5, "-")
operations_with_input(10, 5, "*")
operations_with_input(10, 5, "/")
```

```python
print(operations_with_input_return(10, 5, "+"))
print(operations_with_input_return(10, 5, "-"))
print(operations_with_input_return(10, 5, "*"))
print(operations_with_input_return(10, 5, "/"))
```

```python
def operations_with_match_input(a, b, op):
    match op:
        case "+":
            result = a + b
            print(f"The result of {a} + {b} is: {result}")
        case "-":
            result = a - b
            print(f"The result of {a} - {b} is: {result}")
        case "*":
            result = a * b
            print(f"The result of {a} * {b} is: {result}")
        case "/":
            if b != 0:
                result = a / b
                print(f"The result of {a} / {b} is: {result}")
            else:
                print("Error: Division by zero is not allowed.")
        case _:
            print("Error: Invalid operation. Please choose +, -, *, or /.")
```

```python
def operations_with_match_input_return(a, b, op):
    match op:
        case "+":
            return a + b
        case "-":
            return a - b
        case "*":
            return a * b
        case "/":
            if b != 0:
                return a / b
            else:
                return "Error: Division by zero is not allowed."
        case _:
            return "Error: Invalid operation. Please choose +, -, *, or /."
```

```python
operations_with_match_input(10, 5, "+")
operations_with_match_input(10, 5, "-")
operations_with_match_input(10, 5, "*")
operations_with_match_input(10, 5, "/")
```

```python
print(operations_with_match_input_return(10, 5, "+"))
print(operations_with_match_input_return(10, 5, "-"))
print(operations_with_match_input_return(10, 5, "*"))
print(operations_with_match_input_return(10, 5, "/"))
```

```python
def add_lambda01(a, b):
    return a + b

print(add_lambda01(3, 7))
```

```python
# lambda functions
add_lambda = lambda a, b: a + b
print(add_lambda(3, 7))
```

```python
# lambda functions
add_lambda = lambda a, b: "a is equal to b" if a == b else "a is not equal to b"
print(add_lambda(3, 7))
```

```python
# lambda functions
add_lambda = lambda a, b: a + b
subtract_lambda = lambda a, b: a - b
multiply_lambda = lambda a, b: a * b
```

```python
print(add_lambda(3, 7))
print(subtract_lambda(10, 5))
print(multiply_lambda(4, 6))
```
