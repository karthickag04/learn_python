## 1. Defining and Calling Functions

````python
# Basic function (no parameters)
def greet():
    """Display a greeting message"""
    print("Hello, welcome to Python!")

# Call the function
greet()
greet()  # Can call multiple times
````

````python
# Function with one parameter
def greet_user(name):
    """Greet a specific user"""
    print(f"Hello, {name}!")

greet_user("Alice")
greet_user("Bob")
````

````python
# Function with multiple parameters
def introduce(name, age, city):
    """Introduce a person"""
    print(f"My name is {name}, I'm {age} years old, from {city}.")

introduce("Alice", 25, "New York")
introduce("Bob", 30, "Los Angeles")
````

## 2. Default Parameters

````python
# Function with default values
def greet_custom(name, greeting="Hello"):
    """Greet with custom message"""
    print(f"{greeting}, {name}!")

greet_custom("Alice")              # Uses default greeting
greet_custom("Bob", "Good morning")  # Custom greeting
greet_custom("Charlie", "Hi there")
````

````python
# Multiple default parameters
def create_profile(name, age=18, city="Unknown", country="USA"):
    """Create a user profile"""
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"Location: {city}, {country}")
    print("-" * 20)

create_profile("Alice")  # Only required param
create_profile("Bob", 25)  # Name and age
create_profile("Charlie", 30, "Miami", "USA")  # All params
````

## 3. Keyword Arguments

````python
# Calling with keyword arguments
def describe_pet(animal, name, age):
    """Describe a pet"""
    print(f"I have a {animal} named {name}, {age} years old.")

# Positional arguments (order matters)
describe_pet("dog", "Buddy", 5)

# Keyword arguments (order doesn't matter)
describe_pet(name="Max", age=3, animal="cat")

# Mixed (positional first, then keyword)
describe_pet("rabbit", age=2, name="Fluffy")
````

````python
# Keyword arguments for clarity
def book_flight(origin, destination, date, passengers=1, first_class=False):
    """Book a flight"""
    class_type = "First Class" if first_class else "Economy"
    print(f"\nâœˆï¸ Flight Booking:")
    print(f"   From: {origin} â†’ To: {destination}")
    print(f"   Date: {date}")
    print(f"   Passengers: {passengers}")
    print(f"   Class: {class_type}")

# Clear and readable
book_flight(
    origin="NYC",
    destination="LAX",
    date="2024-06-15",
    passengers=2,
    first_class=True
)
````

## 4. Variable-Length Arguments (*args)

````python
# *args - Accept any number of positional arguments
def add_numbers(*args):
    """Add any number of values"""
    print(f"Received: {args}")
    print(f"Type: {type(args)}")
    return sum(args)

print(f"Sum: {add_numbers(1, 2, 3)}")
print(f"Sum: {add_numbers(10, 20, 30, 40, 50)}")
print(f"Sum: {add_numbers(5)}")
````

````python
# Practical example with *args
def calculate_average(*scores):
    """Calculate average of scores"""
    if not scores:
        return 0
    return sum(scores) / len(scores)

print(f"Average: {calculate_average(85, 90, 78, 92):.1f}")
print(f"Average: {calculate_average(100, 95):.1f}")
````

## 5. Keyword Variable Arguments (**kwargs)

````python
# **kwargs - Accept any number of keyword arguments
def print_info(**kwargs):
    """Print key-value pairs"""
    print(f"Received: {kwargs}")
    print(f"Type: {type(kwargs)}")
    
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

print_info(name="Alice", age=25, city="NYC")
````

````python
# Practical example with **kwargs
def create_user(**user_data):
    """Create user with flexible data"""
    print("\nðŸ‘¤ User Profile:")
    
    # Required fields with defaults
    name = user_data.get("name", "Anonymous")
    email = user_data.get("email", "Not provided")
    
    print(f"   Name: {name}")
    print(f"   Email: {email}")
    
    # Optional fields
    for key, value in user_data.items():
        if key not in ["name", "email"]:
            print(f"   {key.title()}: {value}")

create_user(name="Alice", email="alice@email.com")
create_user(name="Bob", email="bob@email.com", age=30, phone="555-1234")
````

## 6. Combining *args and **kwargs

````python
# Combined - Order matters: regular, *args, **kwargs
def flexible_function(required, *args, **kwargs):
    """Function with all parameter types"""
    print(f"Required: {required}")
    print(f"Args: {args}")
    print(f"Kwargs: {kwargs}")

flexible_function("must have", 1, 2, 3, name="test", value=100)
````

````python
# Real-world example
def log_event(event_type, *messages, **metadata):
    """Log an event with messages and metadata"""
    print(f"\nðŸ“‹ [{event_type.upper()}]")
    
    for msg in messages:
        print(f"   ðŸ“ {msg}")
    
    if metadata:
        print("   ---")
        for key, value in metadata.items():
            print(f"   {key}: {value}")

log_event(
    "INFO",
    "User logged in",
    "Session started",
    user="alice",
    ip="192.168.1.1",
    timestamp="2024-01-15 10:30:00"
)
````

## 7. Docstrings

````python
# Good documentation practice
def calculate_bmi(weight_kg, height_m):
    """
    Calculate Body Mass Index (BMI).
    
    Parameters:
        weight_kg (float): Weight in kilograms
        height_m (float): Height in meters
    
    Returns:
        float: The BMI value
    
    Example:
        >>> calculate_bmi(70, 1.75)
        22.86
    """
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 2)

# Access docstring
print(calculate_bmi.__doc__)

# Or use help()
# help(calculate_bmi)
````

````python
# Test the BMI function
result = calculate_bmi(70, 1.75)
print(f"BMI: {result}")
````

## 8. Complete Example: Order System

````python
def create_order(customer_name, *items, discount=0, **extra_info):
    """
    Create an order with flexible parameters.
    
    Parameters:
        customer_name (str): Name of the customer
        *items: Variable number of (item, price, qty) tuples
        discount (float): Discount percentage (default 0)
        **extra_info: Additional order metadata
    """
    print("=" * 45)
    print("            ORDER RECEIPT")
    print("=" * 45)
    print(f"Customer: {customer_name}")
    
    if extra_info:
        for key, value in extra_info.items():
            print(f"{key.title()}: {value}")
    
    print("-" * 45)
    print(f"{'Item':<20} {'Qty':>5} {'Price':>8} {'Total':>8}")
    print("-" * 45)
    
    subtotal = 0
    for item in items:
        name, price, qty = item
        total = price * qty
        subtotal += total
        print(f"{name:<20} {qty:>5} ${price:>7.2f} ${total:>7.2f}")
    
    print("-" * 45)
    print(f"{'Subtotal:':<35} ${subtotal:>7.2f}")
    
    if discount > 0:
        discount_amount = subtotal * (discount / 100)
        print(f"{'Discount (' + str(discount) + '%):':<35} -${discount_amount:>6.2f}")
        final = subtotal - discount_amount
    else:
        final = subtotal
    
    print("=" * 45)
    print(f"{'TOTAL:':<35} ${final:>7.2f}")
    print("=" * 45)
    
    return final

# Create an order
total = create_order(
    "Alice Johnson",
    ("Laptop", 999.99, 1),
    ("Mouse", 29.99, 2),
    ("USB Cable", 9.99, 3),
    discount=10,
    email="alice@email.com",
    shipping="Express"
)
````

## Summary

### Function Syntax:
```python
def function_name(parameters):
    """Docstring"""
    # code
    return value  # optional
```

### Parameter Types:

| Type | Description | Example |
|------|-------------|---------|
| Positional | Required, order matters | `def f(a, b):` |
| Default | Optional with default value | `def f(a, b=10):` |
| *args | Variable positional | `def f(*args):` |
| **kwargs | Variable keyword | `def f(**kwargs):` |

### Best Practices:
1. Use descriptive function names
2. Add docstrings
3. Keep functions focused (do one thing)
4. Use default parameters wisely
5. Validate inputs when needed
