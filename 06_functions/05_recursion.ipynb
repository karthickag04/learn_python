## 1. Recursion Basics

````python
# Simple countdown
def countdown(n):
    if n <= 0:  # Base case - stop condition
        print("ðŸš€ Liftoff!")
        return
    
    print(n)
    countdown(n - 1)  # Recursive call with smaller problem

countdown(5)
````

````python
# Understanding the call stack
def trace_countdown(n, depth=0):
    indent = "  " * depth
    print(f"{indent}countdown({n}) called")
    
    if n <= 0:
        print(f"{indent}Base case reached, returning")
        return
    
    trace_countdown(n - 1, depth + 1)
    print(f"{indent}countdown({n}) completed")

trace_countdown(3)
````

## 2. Key Components

````python
# Every recursive function needs:
# 1. BASE CASE - When to stop
# 2. RECURSIVE CASE - Call itself with smaller problem
# 3. PROGRESS - Move toward base case

def factorial(n):
    """
    Calculate n! = n Ã— (n-1) Ã— (n-2) Ã— ... Ã— 1
    
    Base case: 0! = 1
    Recursive case: n! = n Ã— (n-1)!
    """
    # Base case
    if n == 0 or n == 1:
        return 1
    
    # Recursive case
    return n * factorial(n - 1)

# Test
for i in range(8):
    print(f"{i}! = {factorial(i)}")
````

````python
# Visualizing factorial
def factorial_verbose(n, depth=0):
    indent = "  " * depth
    print(f"{indent}factorial({n})")
    
    if n == 0 or n == 1:
        print(f"{indent}â†’ returns 1 (base case)")
        return 1
    
    result = n * factorial_verbose(n - 1, depth + 1)
    print(f"{indent}â†’ returns {n} Ã— ... = {result}")
    return result

result = factorial_verbose(4)
````

## 3. Classic Examples

````python
# Fibonacci sequence: 0, 1, 1, 2, 3, 5, 8, 13, 21...
# F(n) = F(n-1) + F(n-2)

def fibonacci(n):
    """Return nth Fibonacci number"""
    if n <= 0:
        return 0
    if n == 1:
        return 1
    
    return fibonacci(n - 1) + fibonacci(n - 2)

# First 12 Fibonacci numbers
print("Fibonacci sequence:")
print([fibonacci(i) for i in range(12)])
````

````python
# Sum of list elements
def sum_list(lst):
    """Recursively sum list elements"""
    if not lst:  # Empty list - base case
        return 0
    
    return lst[0] + sum_list(lst[1:])  # First + rest

numbers = [1, 2, 3, 4, 5]
print(f"Sum of {numbers}: {sum_list(numbers)}")
````

````python
# Count elements in list
def count_elements(lst):
    """Count list elements recursively"""
    if not lst:
        return 0
    return 1 + count_elements(lst[1:])

items = ['a', 'b', 'c', 'd', 'e']
print(f"Count: {count_elements(items)}")
````

````python
# Reverse a string
def reverse_string(s):
    """Reverse string recursively"""
    if len(s) <= 1:
        return s
    return reverse_string(s[1:]) + s[0]

text = "Hello"
print(f"Original: {text}")
print(f"Reversed: {reverse_string(text)}")
````

````python
# Check palindrome
def is_palindrome(s):
    """Check if string is palindrome"""
    s = s.lower().replace(" ", "")  # Clean string
    
    if len(s) <= 1:
        return True
    
    if s[0] != s[-1]:
        return False
    
    return is_palindrome(s[1:-1])

test_words = ["radar", "hello", "racecar", "A man a plan a canal Panama"]
for word in test_words:
    result = "âœ…" if is_palindrome(word) else "âŒ"
    print(f"{result} '{word}'")
````

## 4. Power and GCD

````python
# Power function
def power(base, exp):
    """Calculate base^exp recursively"""
    if exp == 0:
        return 1
    if exp < 0:
        return 1 / power(base, -exp)
    
    return base * power(base, exp - 1)

print(f"2^10 = {power(2, 10)}")
print(f"5^3 = {power(5, 3)}")
print(f"2^-3 = {power(2, -3)}")
````

````python
# Greatest Common Divisor (Euclidean algorithm)
def gcd(a, b):
    """Find GCD using Euclidean algorithm"""
    if b == 0:
        return a
    return gcd(b, a % b)

print(f"GCD(48, 18) = {gcd(48, 18)}")
print(f"GCD(100, 25) = {gcd(100, 25)}")
print(f"GCD(17, 5) = {gcd(17, 5)}")
````

## 5. Recursion with Data Structures

````python
# Flatten nested list
def flatten(lst):
    """Flatten nested list into single list"""
    result = []
    
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))  # Recursive flatten
        else:
            result.append(item)
    
    return result

nested = [1, [2, 3, [4, 5]], 6, [7, [8, [9]]]]
print(f"Nested: {nested}")
print(f"Flat: {flatten(nested)}")
````

````python
# Tree traversal
tree = {
    "value": 1,
    "children": [
        {
            "value": 2,
            "children": [
                {"value": 4, "children": []},
                {"value": 5, "children": []}
            ]
        },
        {
            "value": 3,
            "children": [
                {"value": 6, "children": []}
            ]
        }
    ]
}

def print_tree(node, level=0):
    """Print tree structure"""
    print("  " * level + f"â””â”€ {node['value']}")
    for child in node["children"]:
        print_tree(child, level + 1)

def sum_tree(node):
    """Sum all values in tree"""
    total = node["value"]
    for child in node["children"]:
        total += sum_tree(child)
    return total

print("Tree structure:")
print_tree(tree)
print(f"\nSum of all nodes: {sum_tree(tree)}")
````

## 6. Recursion vs Iteration

````python
# Factorial comparison

# Recursive
def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)

# Iterative
def factorial_iterative(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

n = 10
print(f"Recursive: {n}! = {factorial_recursive(n)}")
print(f"Iterative: {n}! = {factorial_iterative(n)}")
````

````python
# Fibonacci - Inefficient recursive vs efficient iterative
import time

def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

def fib_iterative(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# Test with small number
n = 30
print(f"Computing fib({n})...")

start = time.time()
result1 = fib_recursive(n)
time1 = time.time() - start
print(f"Recursive: {result1} ({time1:.4f}s)")

start = time.time()
result2 = fib_iterative(n)
time2 = time.time() - start
print(f"Iterative: {result2} ({time2:.6f}s)")
````

````python
# Optimized recursion with memoization
from functools import lru_cache

@lru_cache(maxsize=None)
def fib_memoized(n):
    if n <= 1:
        return n
    return fib_memoized(n-1) + fib_memoized(n-2)

n = 100
start = time.time()
result = fib_memoized(n)
elapsed = time.time() - start

print(f"fib({n}) = {result}")
print(f"Time with memoization: {elapsed:.6f}s")
````

## 7. Complete Example: File System Traversal

````python
# Simulated file system
file_system = {
    "name": "root",
    "type": "folder",
    "children": [
        {
            "name": "documents",
            "type": "folder",
            "children": [
                {"name": "report.pdf", "type": "file", "size": 1024},
                {"name": "notes.txt", "type": "file", "size": 256},
                {
                    "name": "projects",
                    "type": "folder",
                    "children": [
                        {"name": "app.py", "type": "file", "size": 2048},
                        {"name": "data.json", "type": "file", "size": 512}
                    ]
                }
            ]
        },
        {
            "name": "images",
            "type": "folder",
            "children": [
                {"name": "photo.jpg", "type": "file", "size": 4096},
                {"name": "logo.png", "type": "file", "size": 1024}
            ]
        },
        {"name": "readme.md", "type": "file", "size": 128}
    ]
}

def print_tree(node, prefix="", is_last=True):
    """Print directory tree"""
    connector = "â””â”€â”€ " if is_last else "â”œâ”€â”€ "
    icon = "ðŸ“" if node["type"] == "folder" else "ðŸ“„"
    
    print(f"{prefix}{connector}{icon} {node['name']}")
    
    if node["type"] == "folder":
        children = node.get("children", [])
        for i, child in enumerate(children):
            new_prefix = prefix + ("    " if is_last else "â”‚   ")
            print_tree(child, new_prefix, i == len(children) - 1)

def calculate_size(node):
    """Calculate total size of folder"""
    if node["type"] == "file":
        return node["size"]
    
    total = 0
    for child in node.get("children", []):
        total += calculate_size(child)
    return total

def find_files(node, extension):
    """Find all files with given extension"""
    results = []
    
    if node["type"] == "file":
        if node["name"].endswith(extension):
            results.append(node["name"])
    else:
        for child in node.get("children", []):
            results.extend(find_files(child, extension))
    
    return results

def count_items(node):
    """Count files and folders"""
    if node["type"] == "file":
        return {"files": 1, "folders": 0}
    
    counts = {"files": 0, "folders": 1}
    for child in node.get("children", []):
        child_counts = count_items(child)
        counts["files"] += child_counts["files"]
        counts["folders"] += child_counts["folders"]
    
    return counts

# Demonstrate
print("ðŸ“‚ FILE SYSTEM")
print("=" * 40)
print_tree(file_system)

print(f"\nðŸ“Š Statistics:")
total_size = calculate_size(file_system)
print(f"   Total size: {total_size:,} bytes")

counts = count_items(file_system)
print(f"   Files: {counts['files']}")
print(f"   Folders: {counts['folders']}")

print(f"\nðŸ” Python files: {find_files(file_system, '.py')}")
print(f"ðŸ” Text files: {find_files(file_system, '.txt')}")
````

## Summary

### Recursion Components:

| Component | Purpose |
|-----------|----------|
| Base case | Stop condition |
| Recursive case | Call itself |
| Progress | Move toward base |

### When to Use:

| Use Recursion | Use Iteration |
|---------------|---------------|
| Tree traversal | Simple loops |
| Graph algorithms | Performance critical |
| Divide and conquer | Limited stack space |
| Elegant solution | Large input |

### Common Pitfalls:
1. Missing base case â†’ infinite recursion
2. Not progressing â†’ stack overflow
3. Redundant calculations â†’ use memoization
