```python
list_of_names = ["ragu", "ravi","raj"]
i = 0
while i < len(list_of_names):
    if(list_of_names[i] == "ravi"):
        print("found ravi")
    
    # print(list_of_names[i])
    i += 1 # i = i + 1
print("*********")
print("*********")
print("*********")
print("*********")

set_of_names = {"ragu", "ravi","raj"}
# set -> convert to list to use index-based while
set_list = list(set_of_names)
idx = 0
while idx < len(set_list):
    if set_list[idx] == "ravi":
        print("found ravi")
    idx += 1

print("*********")
print("*********")
print("*********")
print("*********")


tuple_of_names = ("ragu", "ravi","raj")
# tuple -> indexable
tidx = 0
while tidx < len(tuple_of_names):
    if tuple_of_names[tidx] == "ravi":
        print("found ravi")
    tidx += 1

print("*********")
print("*********")
print("*********")
print("*********")


dict_of_person = {"name": "ragu", "age":20, "location":"Trichy", "othername":"ravi"}
# dict items -> convert to list of (key, value) pairs
items = list(dict_of_person.items())
k = 0
while k < len(items):
    key, value = items[k]
    if value == "ravi":
        print("found ravi")
    k += 1
```
